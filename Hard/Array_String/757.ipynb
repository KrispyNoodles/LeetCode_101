# 757. Set Intersection Size At Least Two

### Problem  
You are given a 2D integer array `intervals` where each interval `intervals[i] = [startᵢ, endᵢ]` represents all integers from `startᵢ` to `endᵢ` inclusive.

A **containing set** is an array `nums` such that **each interval contains at least two integers** from `nums`.

For example, if  
`intervals = [[1,3], [3,7], [8,9]]`,  
valid containing sets include:

- `[1,2,4,7,8,9]`  
- `[2,3,4,8,9]`

Return the **minimum possible size** of such a containing set.

---

### Example 1  
**Input:** `intervals = [[1,3],[3,7],[8,9]]` **Output:** `5`

**Explanation:**  
A valid set is `nums = [2, 3, 4, 8, 9]`.  
No containing set of size 4 exists.

---

### Example 2  
**Input:** `intervals = [[1,3],[1,4],[2,5],[3,5]]` **Output:** `3`

**Explanation:**  
A valid set is `nums = [2, 3, 4]`.  
A size-2 set is impossible.

---

### Example 3  
**Input:** `intervals = [[1,2],[2,3],[2,4],[4,5]]` **Output:** `5`

**Explanation:**  
A valid set is `nums = [1, 2, 3, 4, 5]`.  
Cannot be done with size 4.

---

### Constraints  
- `1 <= intervals.length <= 3000`  
- `intervals[i].length == 2`  
- `0 <= startᵢ < endᵢ <= 10^8`

```python
from typing import List
from itertools import combinations

# my solution
# Brute forcing my way through does not work
class Solution:
    def intersectionSizeTwo(self, intervals: List[List[int]]) -> int:
        
        filtered = []

        # removing arrays that are supersets since we only need the minimum
        # for example when having (2, 9), (3, 5)
        # (2, 9) is not needed as solving (3, 5) solves for it as well
        for current_set in intervals:

            keep = True

            for comparing_set in intervals:

                # if it is a super set of anyset, remove itself
                # creating the set here to use supersets function
                # creating a 'super set' based on boundaries rather than a range set
                if current_set[0] <= comparing_set[0] and current_set[1] >= comparing_set[1]:
                    if current_set != comparing_set:
                        keep = False
                        break
            
            # if after iterating it is not a super set add it to the filtered
            if keep:
                
                # working with the bounds instead of the range
                filtered.append((current_set[0], current_set[1]))


        # sorting the filters from smallest range to biggest (from 24s to 14s amazing)
        filtered.sort(key=lambda x: x[1]-x[0])

        # creating an empty set for all possible values available
        master_set = set()

        # creating a union for all values in there
        for (start, end) in filtered:

            # update is better than union as it doesnt create a set over again
            # the range is only used in here when creating the master set
            master_set.update(range(start, end+1))

        # reordering master_set based on occurence
        # if the value lie in none of the arrays it can be remove
        map = {val: 0 for val in master_set}

        for (start, end) in filtered:
            for value in master_set:
                if start<= value <= end:
                    map[value] +=1

        # sorting the master_set based on the map's value, from big to small
        sorted_master_set = sorted(master_set, key=lambda x: map[x], reverse=True)

        # print(filtered)
        # print(sorted_master_set)
        
        # most you will have is number of arrays * 2 (since each array has 2)
        # min is 2 that solves everything
        for i in range(2, len(filtered) * 2 + 1):

            # combi of that index
            # range of combinations to be created
            # and for all varaitions from (1 to 9)
            combi = combinations(sorted_master_set, i)

            # check the combi here immediately
            # finding the correct answer
            # THIS TAKING THE LONGEST
            for potential_answers in combi:

                # checking how many times it has fulfilled the condition
                # the counter resets whenever a new potential correct_answer is selected
                counter = 0               

                # using left right instead of retrieving it from the array later
                for (left, right) in filtered:

                    count_for_set = 0
                                        
                    # if the intercept of each set is more than or equal 2
                    # it passed for this set and moves on to the next

                    for val in potential_answers:
                        # if the value in combi is within the set's range
                        if left <= val <= right:
                            count_for_set += 1

                            # can skip to the next number in the potential answer as the minimum is 2
                            if count_for_set == 2:
                                counter +=1 
                                break
                    else:
                        break
                
                # if the counter is satisfied, the correct answer has been found
                if counter == len(filtered):
                    
                    # return back the current index as the answer
                    return len(potential_answers)

        return 0
```
