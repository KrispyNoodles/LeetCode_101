# 1047. Remove All Adjacent Duplicates in String

### Problem  
You are given a string `s` consisting of lowercase English letters.  
A **duplicate removal** means selecting **two adjacent and equal letters** and removing them.

You repeatedly perform duplicate removals on `s` until no more can be applied.

Return the **final string** after all removals.  
The answer is guaranteed to be **unique**.

---

### Example 1  
**Input:** `s = "abbaca"` **Output:** `"ca"`

**Explanation:**  
- `"bb"` is removed → `"aaca"`  
- `"aa"` is removed → `"ca"`

---

### Example 2  
**Input:** `s = "azxxzy"` **Output:** `"ay"`

---

### Constraints  
- `1 <= s.length <= 10^5`  
- `s` consists only of lowercase English letters  


# Exploration

```python
class Solution:
    def removeDuplicates(self, s: str) -> str:
        
        # stack to contain dup
        remaining_stack = []

        for index, char in enumerate(s):

            # adding the values into the stack the stack if it is empty
            if not remaining_stack:
                remaining_stack.append(char)
                # skip this loop
                continue
            
            # if the current value is the same as the previous found in the stack
            # remove it from the stack and skip
            if char == remaining_stack[-1]:

                # remove it from the stack
                remaining_stack.pop()
            
            # if it is not duplicated, add it to the stack
            else:
                remaining_stack.append(char)

        # reconstrcut the stack
        return "".join(remaining_stack)
        
        
```

```python
sol = Solution()

# Input: 
s = "abbaca"
# Output: "ca"

# Input: 
s = "azxxzy"
# Output: "ay"

sol.removeDuplicates(s)
```
