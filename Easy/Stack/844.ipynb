# 844. Backspace String Compare

### Problem  
You are given two strings `s` and `t`.  
Return **true** if both strings are equal **after simulating typing into a text editor**.

`#` represents a **backspace**:

- If `#` is typed when the text is empty, it stays empty  
- Otherwise, the previous character is removed  

---

### Example 1  
**Input:** `s = "ab#c", t = "ad#c"` **Output:** `true`  
**Explanation:** Both evaluate to `"ac"`.

---

### Example 2  
**Input:** `s = "ab##", t = "c#d#"` **Output:** `true`  
**Explanation:** Both evaluate to an empty string `""`.

---

### Example 3  
**Input:** `s = "a#c", t = "b"` **Output:** `false`  
**Explanation:** `s → "c"` while `t → "b"`.

---

### Constraints  
- `1 <= s.length, t.length <= 200`  
- Strings contain only lowercase letters and `#`  


# Exploration

```python
class Solution:
    def backspaceCompare(self, s: str, t: str) -> bool:

        stack_1 = []
        stack_2 = []

        for char in s:

            # if hashtag means that it is deleting the last inserted elemt
            if char == '#':
                
                # just remove the last element
                # checking is the stack is empty (if not nothing is done)
                if stack_1:
                    stack_1.pop()
            else:
                stack_1.append(char)

        for char in t:

            # if hashtag means that it is deleting the last inserted elemt
            if char == '#':

                # checking is the stack is empty (if not nothing is done)
                if stack_2:
                    stack_2.pop()
            else:
                stack_2.append(char)
        
        if stack_1 == stack_2:
            return True

        else:
            return False
        
```

```python
sol = Solution()
s = "a##c"

t = "#a#c"

sol.backspaceCompare(s,t)
# Output: true
# Explanation: Both s and t become "ac".
```
