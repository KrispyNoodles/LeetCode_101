```python
class MyQueue:

    def __init__(self):
        # initialising an empty stack
        self.num_stack = []

    # pushing a number into the stack
    def push(self, x: int) -> None:
        self.num_stack.append(x)

    # the stack is reverse
    def pop(self) -> int:
        
        value = self.num_stack[0]

        # creating a new slice starting (from index 1 to the end)
        self.num_stack = self.num_stack[1:]

        return value
    
    # the peek returns the first value on the left
    def peek(self) -> int:
        return self.num_stack[0]

    def empty(self) -> bool:
        # return true if num_stack is empty
        # else return false

        # therefore it is a (not) false for empty stack
        return not self.num_stack
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
