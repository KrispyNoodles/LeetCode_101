# 225. Implement Stack Using Queues

### Problem  
Implement a **Last-In-First-Out (LIFO)** stack using **only two queues**.

Implement the `MyStack` class with the following operations:

- **void push(int x)** — push element `x` onto the stack  
- **int pop()** — remove and return the top element  
- **int top()** — return the top element  
- **boolean empty()** — return `true` if the stack is empty, otherwise `false`

#### Notes:
- You may only use **standard queue operations**:  
  push to back, peek/pop from front, check size, check empty  
- If your language doesn’t have a queue, you may simulate it using a list or deque, but still must obey queue rules.

---

### Example 1  
**Input:**  
`["MyStack", "push", "push", "top", "pop", "empty"]`  
`[[], [1], [2], [], [], []]`  
**Output:**  
`[null, null, null, 2, 2, false]`

**Explanation:**  
```
MyStack myStack = new MyStack();
myStack.push(1);
myStack.push(2);
myStack.top();   // returns 2
myStack.pop();   // returns 2
myStack.empty(); // returns false
```

````python
class MyQueue:

    def __init__(self):
        # initialising an empty stack
        self.num_stack = []

    # pushing a number into the stack
    def push(self, x: int) -> None:
        self.num_stack.append(x)

    # the stack is reverse
    def pop(self) -> int:
        
        value = self.num_stack[0]

        # creating a new slice starting (from index 1 to the end)
        self.num_stack = self.num_stack[1:]

        return value
    
    # the peek returns the first value on the left
    def peek(self) -> int:
        return self.num_stack[0]

    def empty(self) -> bool:
        # return true if num_stack is empty
        # else return false

        # therefore it is a (not) false for empty stack
        return not self.num_stack
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
````
