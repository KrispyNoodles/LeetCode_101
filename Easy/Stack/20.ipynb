```python
# using a stack and a map
class Solution:
    def isValid(self, s: str) -> bool:
        
        # correct map
        mapping_values = {'(':')', '[':']', '{':'}'}

        # array that keeps only open brackets
        open_brack_stack = []

        for char in s:
            # it only find the keys, which are the close brackets
            if char in mapping_values.keys():
                open_brack_stack.append(char)

            # it only find the keys, which are the open brackets
            elif char in mapping_values.values():

                # if it is empty, it means there was no open brackets appended previously
                if open_brack_stack == []:
                    return False
                
                # this means that there isnt a mtach of correct brackets
                # ensuring the open bracket's key maps to the correct close brackets
                if char != mapping_values[open_brack_stack.pop()]:
                    return False
        
        # checking if open_brack_stack is empty in the end
        return not open_brack_stack
```

```python
# using another function that checks for curent and last
# this relies alot on the final len of the empty stack to check if the brackets were opened and closed correclty
class Solution:
    def isValid(self, s: str) -> bool:
        
        empty_stack = []

        for i in range(len(s)):
           
            # if stack exists
            if empty_stack:

                # last of the stack is
                last = empty_stack[-1]

                # trying to find the pair when encountering a open
                if self.is_pair(last, s[i]):

                    # removed after the pair has been found
                    empty_stack.pop()

                    # move on to the next loop without appending
                    continue
            
            empty_stack.append(s[i])

        # if the stack isnt empty it means the is_pair return false which caused it not remove that pair
        return not empty_stack
                                
    
    def is_pair(self, last, current):
        if last == '(' and current == ')':
            return True

        elif last == '{' and current == '}':
            return True
        
        elif last == '[' and current == ']':
            return True
        
        else:
            return False
```

```python
s = "(("
# False

sol = Solution()

# # Input: 
s = "({{{{}}}))"
# # Output: False

# # Input: 
# s = "()))"
# # Output: False

# # Input: 
s = "()[]{}"
# # Output: True

# # Input: 
s = "){"
# # Output: False

# # Input: 
s = "(){]())"
# # Output: True

sol.isValid(s)
```
