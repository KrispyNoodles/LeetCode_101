# 682. Baseball Game

### Problem  
You are keeping score for a baseball game using unusual rules.  
You begin with an **empty record**.

You are given an array `operations`, where each element is one operation to apply:

- **Integer `x`** → record a new score of `x`  
- **"+"** → record the sum of the previous **two** scores  
- **"D"** → record **double** the previous score  
- **"C"** → remove the **previous** score  

Return the **sum of all recorded scores** after applying all operations.

All operations are valid, and all values fit in a 32-bit integer.

---

### Example 1  
**Input:** `ops = ["5","2","C","D","+"]` **Output:** `30`

**Explanation:**  
```
"5" → [5]  
"2" → [5, 2]  
"C" → [5]  
"D" → [5, 10]  
"+" → [5, 10, 15]  
Total = 30
```

---

### Example 2  
**Input:** `ops = ["5","-2","4","C","D","9","+","+"]` **Output:** `27`

---

### Example 3  
**Input:** `ops = ["1","C"]` **Output:** `0`

---

### Constraints  
- `1 <= operations.length <= 1000`  
- Each operation is `"C"`, `"D"`, `"+"`, or an integer string in the range `[-3*10^4, 3*10^4]`  
- `"+"` → always at least **two** previous scores  
- `"C"` and `"D"` → always at least **one** previous score  


# EXploration

````python
from typing import List

# my solution
class Solution:
    def calPoints(self, operations: List[str]) -> int:
        # print(operations)

        stack_score = []

        for op in operations:

            # if operation found
            # pop operation on the scores
            if op == 'C':

                # removing the most recent score
                stack_score.pop()
            
            # duplicate operation
            elif op == 'D':
                
                # duplicate the last number added the stack_score
                # then append back
                score_to_be_added = stack_score[-1] * 2
                stack_score.append(score_to_be_added)
            
            elif op == '+':

                # add the last and second last number of the stack_score
                # then append back
                score_to_be_added = stack_score[-1] + stack_score[-2]
                stack_score.append(score_to_be_added)

            # it is a number detected
            else:
                stack_score.append(int(op))

        # return the sum
        return sum(stack_score)
````

````python
s = Solution()
# Input: 
ops = ["5","2","C","D","+"]

s.calPoints(ops)

# Output: 30
# Explanation:
# "5" - Add 5 to the record, record is now [5].
# "2" - Add 2 to the record, record is now [5, 2].
# "C" - Invalidate and remove the previous score, record is now [5].
# "D" - Add 2 * 5 = 10 to the record, record is now [5, 10].
# "+" - Add 5 + 10 = 15 to the record, record is now [5, 10, 15].
# The total sum is 5 + 10 + 15 = 30.
````
