```python
from typing import List

# my solution
class Solution:
    def calPoints(self, operations: List[str]) -> int:
        # print(operations)

        stack_score = []

        for op in operations:

            # if operation found
            # pop operation on the scores
            if op == 'C':

                # removing the most recent score
                stack_score.pop()
            
            # duplicate operation
            elif op == 'D':
                
                # duplicate the last number added the stack_score
                # then append back
                score_to_be_added = stack_score[-1] * 2
                stack_score.append(score_to_be_added)
            
            elif op == '+':

                # add the last and second last number of the stack_score
                # then append back
                score_to_be_added = stack_score[-1] + stack_score[-2]
                stack_score.append(score_to_be_added)

            # it is a number detected
            else:
                stack_score.append(int(op))

        # return the sum
        return sum(stack_score)
```

```python
s = Solution()
# Input: 
ops = ["5","2","C","D","+"]

s.calPoints(ops)

# Output: 30
# Explanation:
# "5" - Add 5 to the record, record is now [5].
# "2" - Add 2 to the record, record is now [5, 2].
# "C" - Invalidate and remove the previous score, record is now [5].
# "D" - Add 2 * 5 = 10 to the record, record is now [5, 10].
# "+" - Add 5 + 10 = 15 to the record, record is now [5, 10, 15].
# The total sum is 5 + 10 + 15 = 30.
```
