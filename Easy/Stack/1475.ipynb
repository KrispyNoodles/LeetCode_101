```python
# creating increasing monotnic stack
prices = [8,4,6,2,3]

# FINALLYL CORRECT INCREASING LOLS

# stack contains index rather values
val_stack = []

# final_ans array
final_ans = [0]*len(prices)

for index in range(len(prices)):

    # checking if stack exists
    # if the current element is more
    # AND the condition that the current value is higher than the one in the stack

        # it will then remove the index
    while val_stack and prices[index] >= prices[val_stack[-1]]:

        # the current's index next highest value is...
        pop_index = val_stack.pop()

        # the pop's value's highest value is the one current encountered
        final_ans[pop_index] = prices[index]
    
    val_stack.append(index)

# val_stack will contain all those that remains the highest for itself
for index in val_stack:
    final_ans[index] = prices[index]

final_ans
```

```python
from typing import List

class Solution:
    def finalPrices(self, prices: List[int]) -> List[int]:

        # I am finidng the next lowest
        # so this is a decreasing monotonic stack

        # stack contains index rather values

        # TRY AS MUCH AS POSSIBLE TO USE INDEXES IN THE STACK RATHER THAN THE ACTUAL VALUES ITSELF
        # AS THE VALUES CAN BE RETREIEVED BACK BY MAPPING THE INDEX IN THE STACK TO THE ORIGINAL ARRAY

        val_stack = []

        # final_ans array
        discount_arr = [0]*len(prices)

        for index in range(len(prices)):

            # checking if stack exists
            # if the current element is more
            # AND the condition that the current value is higher than the one in the stack

                # it will then remove the index
            while val_stack and prices[index] <= prices[val_stack[-1]]:

                # the current's index next highest value is...
                pop_index = val_stack.pop()

                # the pop's value's highest value is the one current encountered
                discount_arr[pop_index] = prices[index]
            
            val_stack.append(index)

        # val_stack will contain all those that remains the highest for itself
        # these are prices that will not have any discount
        for index in val_stack:
            discount_arr[index] = 0

        final_ans = []

        # checking for negative
        for index in range(len(prices)):
            final_cost = prices[index] - discount_arr[index]

            final_ans.append(final_cost)


        return final_ans
```

```python
s = Solution()
prices = [10,1,1,6]


s.finalPrices(prices)
# Output: [9,0,1,6]

```
