```python
from typing import List

# my solution however I am not optimising on using the stack 
class Solution:
    def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:

        # create a stack to append and pop for comparison
        val_stack = []

        # for each element in nums1 it is compared to every in nums2
        for value in nums1:

            # finding the position
            position = nums2.index(value)
            
            temp_array = nums2[position+1:]

            # if the array is empty the element is the largetst
            # so append -1
            if not temp_array:
                val_stack.append(-1)
                continue

            # assume the value is the biggest
            val_stack.append(value)

            # for values in temp_array do comparison
            for i in temp_array:                

                # pop it if it is the smallest
                # and append the bigger value instead
                # it can break as it wants to find it asap
                if value < i:
                    val_stack[-1] = i
                    break

                # the current value is the biggest and add '-1' instead
                # do not break as it needs to check through all elements
                elif val_stack[-1] == value:
                    val_stack[-1] = -1

        return val_stack
        
```

```python
# Input: 
nums1 = [4,1,2]
nums2 = [1,3,4,2]
# Output: [-1,3,-1]

# nums1 = [4,1,2]
# nums2 = [1,2,3,4]
# Output: [-1, 2, 3]

# nums1 =[1,3,5,2,4]
# nums2 = [6,5,4,3,2,1,7]
# Output: [7,7,7,7,7]


s = Solution()
s.nextGreaterElement(nums1, nums2)

```
