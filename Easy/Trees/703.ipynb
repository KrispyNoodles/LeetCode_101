# 703. Kth Largest Element in a Stream

```python
from typing import List

class KthLargest:

    def __init__(self, k: int, nums: List[int]):

        nums.sort(reverse=True)

        # array containing the scores
        self.array = nums

        # top highest how many to ekeep track
        self.k = k
        

    def add(self, val: int) -> int:

        self.array.append(val)

        # sort
        self.array.sort(reverse=True)

        # return the largest value
        return self.array[self.k-1]

# Your KthLargest object will be instantiated and called as such:
# obj = KthLargest(k, nums)
# param_1 = obj.add(val)

# time complextiy of O(nlog(n))
```

```python
from typing import List

# using insertion sort
class KthLargest:

    def __init__(self, k: int, nums: List[int]):

        # array containing the scores
        self.array = self.insertion_sort(nums)

        # top highest how many to ekeep track
        self.k = k        

    def add(self, val: int) -> int:

        # add the value in
        self.array.append(val)

        # sort
        self.array = self.insertion_sort(self.array[:])
        
        # return the largest value
        return self.array[self.k-1]

    def insertion_sort(self, array):

        for index in range(1, len(array)):

            j = index-1

            # if the value ios more then it has to be swapped and shifted
            while j >= 0 and array[j+1] > array[j]:

                temp_val = array[j+1]
                array[j+1] = array[j]
                array[j] = temp_val

                j-=1

        return array        

# Your KthLargest object will be instantiated and called as such:
# obj = KthLargest(k, nums)
# param_1 = obj.add(val)

# time complextiy of O(nlog(n))
```
