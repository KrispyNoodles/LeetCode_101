# 1523. Count Odd Numbers in an Interval Range

```python
# brute force
# TLE (time limit exceeded)

class Solution:
    def countOdds(self, low: int, high: int) -> int:

        # creating a list from low to high

        list = range(low, high+1)

        # counter
        counter = 0

        for values in list:
            if values %2 != 0:
                counter+=1

        return counter
```

```python
class Solution:
    def countOdds(self, low: int, high: int) -> int:

        # calculate the difference the high and low
        # this returns the number of values in the range
        difference = high+1-low

        # checking for high == low
        # just check for one value
        if difference == 1:
            if high%2 == 0:
                return 0
            else:
                return 1
        
        # if len is odd and high is odd (+1)
        elif difference%2 != 0 and high%2 != 0:
            return difference//2 + 1

        # if len is odd and high is even (-1)
        # have to include + 1 because of how odd numbers are being divided
        # this is just the same as the case below
        elif difference%2 != 0 and high%2 == 0:
            return difference//2 + 1 - 1

        else:
            return difference//2
```

```python
# optimised
class Solution:
    def countOdds(self, low: int, high: int) -> int:

        # calculate the difference the high and low
        # this returns the number of values in the range
        difference = high+1-low

        # checking for high == low
        # just check for one value
        if difference == 1:
            if high%2 == 0:
                return 0
            else:
                return 1
        
        # if len is odd and high is odd (+1)
        elif difference%2 != 0 and high%2 != 0:
            return difference//2 + 1

        else:
            return difference//2
```

```python
low = 8
high = 10

sol = Solution()
sol.countOdds(low, high)

# Output: 1
```
