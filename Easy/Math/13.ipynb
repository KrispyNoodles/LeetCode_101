```python
# optimised

class Solution:
    def romanToInt(self, s: str) -> int:
        
        # creating a dict
        roman_dict = {"I": 1, "V": 5, "X":10, "L":50, "C":100, "D":500, "M":1000}

        # special cases
        set_special = {"I","X","C"}
        special_dict = {"IV": 4, "IX": 9, "XL":40, "XC":90, "CD":400, "CM":900}

        # counter and index
        sum = 0
        index = 0

        while index < len(s):

            if s[index] in set_special and index+1 < len(s):

                new_special_char = s[index] + s[index+1]

                if new_special_char in special_dict:
                    sum += special_dict[new_special_char]
                    index += 2 

                # count normally
                else:
                    sum = roman_dict[s[index]] + sum
                    index+=1         

            else:
                sum = roman_dict[s[index]] + sum
                index+=1

        return sum
    
# time complexity of O(N)
# space complecity of O(1) because of the fixed values
```

```python
sol = Solution()

# Input: 
s = "MCMXCIV"

# Output: 1994
# Explanation: MCMXCIV = 1994
sol.romanToInt(s)
```
