# 1925. Count Square Sum Triples

```python
# recap combination
from itertools import permutations

arra = [1,2,3]

new = permutations(arra, 2)

for pairs in new:
    print(pairs)
```

```python
class Solution:
    def countTriples(self, n: int) -> int:
        
        # creating an array of the squares
        square_array = []

        for values in range(1, n+1):

            square_array.append(values**2)

        # creating permutations of pair
        pair_permutations = permutations(square_array, 2)

        answer_val = 0

        # creating a value for the max
        max_val = max(square_array)

        # adding both pairs together, if more than max then move next
        for pairs in pair_permutations:
            
            sum_value = pairs[0] + pairs[1]

            # if the combined value is more than the max value in the array, then continue
            if sum_value > max_val:
                continue

            elif sum_value in square_array:
                answer_val += 1

        return answer_val

# time complexity of O(n) because of pair_permutations loop and because of checking the value IN the array
# space complexity of O(n^2) because of pair_permutations
```

```python
# solving without permutations

class Solution:
    def countTriples(self, n: int) -> int:
        
        # creating an array of the squares
        square_array = []

        for values in range(1, n+1):

            square_array.append(values**2)

        answer_val = 0

        # covnerting to set so then when using finding it, it uses O(1) instead of O(n)
        square_set = set(square_array)

        # creating a value for the max
        max_val = max(square_array)
        
        # adding both pairs together, if more than max then move next
        for index_1 in range(len(square_array)):
            temp_arr = square_array[:]
            
            temp_arr.pop(index_1)
            
            for index_2 in range(len(temp_arr)):
                sum_value = square_array[index_1] + temp_arr[index_2]

                # if the combined value is more than the max value in the array, then continue
                if sum_value > max_val:
                    continue

                elif sum_value in square_set:
                    answer_val += 1

        return answer_val

# time complexity of O(n^2) because of running 2 for loops
# space complexity of O(n) because of pair_permutations
```

```python
# Input: 
n = 5

sol = Solution()
sol.countTriples(n)

# Output: 2
# Explanation: The square triples are (3,4,5) and (4,3,5).
```
