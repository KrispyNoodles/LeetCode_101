```python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val

        # ah it points not to a value but to a node
        self.next = next

# 3->2->1
# become 3<-2<-1
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        
        # assigning a variable where current is (only done once)
        # its like an initalisation of the loop
        # it is a pointer not a duplicate of variable
        current = head
        prev = None
 
        # prints the current position
        print(f"this gets printed once {current.val}")


        # as long as there is a current assigned
        while current:

            # # gets reasign, and loop through again
            next = current.next

            # # reassiging the next of the first var to be
            current.next = prev
            prev = current

            current = next

            print(current.val)
        return prev
```

```python
sol = Solution()
a = ListNode(1,ListNode(2,ListNode(3,ListNode(4,ListNode(5)))))

sol.reverseList(a)
```
