# 944. Delete Columns to Make Sorted

```python
from typing import List

class Solution:
    def minDeletionSize(self, strs: List[str]) -> int:
        
        # checkingg the len of string
        if strs:
            len_of_string = len(strs[0])
        else:
            return 0

        counter = 0

        for index in range(len_of_string):

            temp_char = ''

            for strings in strs:

                chracter = strings[index]

                # if it is the first temp_char then temp_char = character
                if not temp_char:
                    temp_char = chracter

                # if the temp_char is lesser than continue and add the counter
                if temp_char > chracter:
                    counter += 1
                    break
                else:
                    temp_char = chracter

        return counter
```
