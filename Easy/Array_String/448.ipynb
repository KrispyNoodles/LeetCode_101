# 448. Find All Numbers Disappeared in an Array

# Exploration

```python
# THE NATURE OF HOW RANGE WORKS

# range(start, end) includes the start value but stops before the end value
# range(a, b) creates numbers from: a â†’ b-1
# therefore range(1, 4) will generate 1, 2, 3
perf_set = set(range(1, 4))

print(perf_set)
```

```python
# however when range is used with the len of an array it starts from 0 to the (length-1)

list_1 = [1, 2, 3]
range_1 = range(len(list_1))

for i in range_1:
    print(i)
```

```python
from typing import List

# solution 1 using sets
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:

        # creating a perfect set from the len
        print(len(nums))

        # len(nums) + 1 because the nature of how range works
        perf_set = set(range(1, len(nums)+1))

        print(perf_set)
        
        # converting it into a set
        missing = perf_set - set(nums)
        
        # retrieving the values into an array
        final_array = []

        for i in missing:
            final_array.append(i)

        return final_array
```

```python
s = Solution()

nums = [4,3,2,7,8,2,3,1]
# Output: [5,6]

nums = [1,1]
# Output: [2]

s.findDisappearedNumbers(nums)
```

```python
from typing import List

# solution 2 using map
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:

        # creating a dict of an ideal range
        # map = {}

        # for i in range(1, len(nums)+1):
        #     map[i] = 0

        # alteranatively for creating a map
        map = {i: 0 for i in range(1, len(nums)+1)}

        # countign the number times the values appear into the map
        for val in nums:
            map[val] += 1

        final_array = []

        # iterating through the map to retrieve the values that did not appear
        for key, value in map.items():
            if value == 0:
                final_array.append(key)

        return final_array
```

```python
s = Solution()

nums = [4,3,2,7,8,2,3,1]
# Output: [5,6]

# nums = [1,1]
# Output: [2]

s.findDisappearedNumbers(nums)
```

```python
from typing import List

# solution 2 using map
class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        
        # creating an array of 0
        new_array = [0] * len(nums)
        

        # for each value's index in the current list , include a -1 for each time it appears        
        for val in nums:
            new_array[val-1] = -1

        # retrieivng the missing values
        # final_array = []
        
        # for index, value in enumerate(new_array):
        #     print(index)
        #     if value == 0:
        #         final_array.append(index+1)

        # return final_array

        # this works the same (but alot faster because there is no variable to be created)
        return [index + 1 for index, value in enumerate(new_array) if value==0]
```

```python
s = Solution()

nums = [4,3,2,7,8,2,3,1]
# Output: [5,6]

# nums = [1,1]
# Output: [2]

s.findDisappearedNumbers(nums)
```
