```python
# my sol
class Solution(object):
    def twoSum(self, nums, target):

        # indexes of first and secodn value
        first_value = 0
        second_value = 0

        # target has two numbers
        for i in range(len(nums)):
            
            # for each value (first_value) in the list
            # use the target - the first_value and then find the second value in the list
            value_2 = target - nums[i]

            # creating a temp copy
            temp_list = nums[:]
            
            # creating a temp list and removing the current target from it
            # pop removes based off position
            temp_list.pop(i)

            # seeing if the value is found
            if value_2 in temp_list:
                first_value = i
                break
            
        # finding the second_value's index
        for i in range(len(nums)):
            
            # if the value that was subtracted is match, break from the loop
            if value_2 == nums[i] and i != first_value:
                second_value = i
                break 
        
        return [first_value, second_value]
```

```python
# Approach 1: Brute Force

class Solution:
    def twoSum(self, nums, target):
        for i in range(len(nums)):

            # it knows that the first value of the array is no longer needed in the solution, therefore the range is shifted
            for j in range(i+1, len(nums)):
                
                # if the second value is matched
                if nums[j] == target-nums[i]:
                    return [i, j]
                
        # after running through if there is no soluitioin
        return []
```

time complexity: $O(n^2)$ because there are two for loops \
space complexity: $O(1)$ because the space does not depend on the size of the input array

```python
# Approach 2: Two-pass Hash Table (Similar to my code actually)

class Solution:
    def twoSum(self, nums, target):

        hashmap = {}

        # creating the hashmap
        for i in range(len(nums)):
            
            # the value of the range, mapping to the index
            hashmap[nums[i]] = i

        # finidn the complement
        for i in range(len(nums)):
            complement = target - nums[i]

            # if the value of the complement is found in the hashmap and its index is not equivalent to the current point
            # then the pair is found
            if complement in hashmap and hashmap[complement] != i:
                return [i, hashmap[complement]]
            
        # after running through if there is no soluitioin
        return []
    
```

time complexity: $O(n)$ because we traversthelit contianing the n eelmenth only twice \
space complexity: $O(n)$ because the space required depends on the number of items stored in the hash table

```python
# Approach 3: One-pass Hash Table

class Solution:
    def twoSum(self, nums, target):

        hashmap = {}

        for i in range(len(nums)):

            complement = target - nums[i]

            # while we are iterating through the list we can look back to check if the current element's complement is already inside
            if complement in hashmap:
                return [i, hashmap[complement]]
            
            hashmap[nums[i]] = i

        # after running through if there is no soluitioin
        return []
    
```

time complexity: $O(n)$ because we travers the list contianing the n elements only once \
space complexity: $O(n)$ because the space required depends on the number of items stored in the hash table
