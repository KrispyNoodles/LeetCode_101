# 1768. Merge Strings Alternately

### Problem  
You are given two strings `word1` and `word2`.  
Merge the strings by adding letters in **alternating order**, starting with `word1`.

If one string is longer, append the **remaining letters** at the end.  
Return the merged string.

---

### Example 1
**Input:** `word1 = "abc", word2 = "pqr"` **Output:** `"apbqcr"`

**Explanation:**  
```
word1:  a   b   c
word2:    p   q   r
merged: a p b q c r
```

---

### Example 2
**Input:** `word1 = "ab", word2 = "pqrs"` **Output:** `"apbqrs"`

**Explanation:**  
```
word1:  a   b 
word2:    p   q   r   s
merged: a p b q   r   s
```

---

### Example 3
**Input:** `word1 = "abcd", word2 = "pq"` **Output:** `"apbqcd"`

**Explanation:**  
```
word1:  a   b   c   d
word2:    p   q 
merged: a p b q c   d
```

---

### Constraints
- `1 <= word1.length, word2.length <= 100`  
- `word1` and `word2` consist of lowercase English letters


# Exploration

````python
# converting an array into a string
array = ['a', 'b', 'c']
string = "".join(array)
string
````

````python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        
        # checking which is longer first
        len_1 = len(word1)
        len_2 = len(word2)

        max_val = max(len_1, len_2)
        min_val = min(len_1, len_2)

        for _ in range(max_val-min_val):
            
            # word 1 needs to be appended
            if min_val == len_1:
                # appending empty strings
                word1 = word1 + "_"
            
            if min_val == len_2:               
                # appending empty strings
                word2 = word2 + "_"

        final_array = []

        for int in range(max_val):

            if word1[int] == "_":
                final_array.append(word2[int])
                # continue iteration
                continue

            elif word2[int] == "_":
                final_array.append(word1[int])
                continue

            else:
                final_array.append(word1[int])
                final_array.append(word2[int])
                
        return ("".join(final_array))
````

````python
word1 = "ab"
word2 = "pqrs"

s = Solution()
s.mergeAlternately(word1, word2)

# Output: "apbqcr"
````

````python
# Approach 1: Two Pointers
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:

        m = len(word1)
        n = len(word2)
        i = 0
        j = 0

        result = []

        # the loop increments as long as the counters as lesser than the len of the words
        while i < m or j < n:
            
            # incremmenet as long as it is still lesser than the len
            # once the value incrementted is too large, it will fall into the other loop
            if i < m:
                # append word
                result += word1[i]

                # increment counter
                i += 1
            
            if j < n:
                # append word
                result += word2[j]
            
                # increment counter
                j += 1
            
        return "".join(result)

````

````python
word1 = "ab"
word2 = "pqrs"

s = Solution()
s.mergeAlternately(word1, word2)

# Output: "apbqcr"
````

time complexity: $O(m + n)$ m and n represents len of word 1 and word 2, we iterate over word 1 and word 2 once and push their letters into result \
space complexity: $O(1)$ Without considering the space consumed by the input strings (word1 and word2) and the output string (result), we do not use more than constant space

````python
# Approach 2: One Pointer
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:

        result = []

        # max value
        n = max(len(word1), len(word2))

        for i in range(n):

            if i < len(word1):
                result += word1[i]
            
            if i < len(word2):
                result += word2[i]
            
        return "".join(result)

````

````python
word1 = "ab"
word2 = "pqrs"

s = Solution()
s.mergeAlternately(word1, word2)

# Output: "apbqcr"
````

Time complexity: $O(m+n)$ We iterate over word1 and word2 once pushing their letters into result. It would take O(m+n) time. \
Space complexity: $O(1)$ Without considering the space consumed by the input strings (word1 and word2) and the output string (result), we do not use more than constant space.
