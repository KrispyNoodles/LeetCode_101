# 217. Contains Duplicate

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        if len(set(nums))<len(nums):
            return True

        else:
            return False

# O(n) because for each time the lenght of the nums increase, the complexity increases too
```

```python
s = Solution()
nums = [1,2,3,1]
# True

s.hasDuplicate(nums)
```
