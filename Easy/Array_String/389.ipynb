```python
# however this does not account for same letters
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:

        # sort first then find
        s = sorted(s)
        t = sorted(t)

        for index in range(len(s)):
            
            if t[index] != s[index]:
                return t[index]

        # if it succeeds means it is at the end
        return t[-1]

# time complexity of O(nlog(n))
```

```python
# cheese solution using sum

# however this does not account for same letters
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:

        # ord function converts char into asci value
        # unicode code of a specified character.
        sum_t = 0

        for char in t:
            sum_t += ord(char)

        sum_s = 0

        for char in s:
            sum_s += ord(char)

        difference = sum_t - sum_s

        # chr convers back from ascii value to character
        return chr(difference)
```
