# 724. Find Pivot Index

### Problem  
Given an integer array `nums`, compute the **pivot index** of the array.

The pivot index is an index where:

- the **sum of all elements strictly to the left** equals  
- the **sum of all elements strictly to the right**

If the index is at the far left, the left sum is `0`.  
If at the far right, the right sum is `0`.

Return the **leftmost** pivot index.  
If none exists, return `-1`.

---

### Example 1  
**Input:** `nums = [1,7,3,6,5,6]` **Output:** `3`

**Explanation:**  
Left sum = `1 + 7 + 3 = 11`  
Right sum = `5 + 6 = 11`

---

### Example 2  
**Input:** `nums = [1,2,3]` **Output:** `-1`

---

### Example 3  
**Input:** `nums = [2,1,-1]` **Output:** `0`

**Explanation:**  
Left sum = `0`  
Right sum = `1 + (-1) = 0`

---

### Constraints  
- `1 <= nums.length <= 10^4`  
- `-1000 <= nums[i] <= 1000`

```python
for i in range(len(nums)):

    # if i is first
    if i==0:
        left_sum = nums[0]
        (L2, R2) = (1, len(nums)-1)
        right_sum = new_array[R1+1] - new_array[L1]

        if left_sum == right_sum:
            print(f"answer_found at {i}")
    
    # if i is last
    if i == (len(nums)-1):
        (L1, R1) = (0, len(nums)-2)
        left_sum = new_array[R1+1] - new_array[L1]
        right_sum = nums[-1]

        if left_sum == right_sum:
            print(f"answer_found at {i}")
    
    # the bounds are moving
    else:
        (L1, R1) = (0, i-1)
        (L2, R2) = (i+1, len(nums)-1)
        left_sum = new_array[R1+1] - new_array[L1]
        right_sum = new_array[R2+1] - new_array[L2]

        if left_sum == right_sum:
            print(f"answer_found at {i}")
```

```python
class Solution(object):
    def pivotIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # new array containing sum
        new_array = [0]
        sum_total = 0

        # total_sum is the value on the far right

        for val in nums:
            sum_total = sum_total + val
            new_array.append(sum_total)

        for i in range(len(nums)):

            # if i is first
            if i == 0:
                left_sum = 0
                (L2, R2) = (1, len(nums)-1)

                right_sum = new_array[R2+1] - new_array[L2]

                if left_sum == right_sum:
                    return i
            
            # if i is last
            if i == (len(nums)-1):
                (L1, R1) = (0, len(nums)-2)
                left_sum = new_array[R1+1] - new_array[L1]
                right_sum = 0

                if left_sum == right_sum:
                    return i
            
            # the bounds are moving
            else:
                (L1, R1) = (0, i-1)
                (L2, R2) = (i+1, len(nums)-1)
                left_sum = new_array[R1+1] - new_array[L1]
                right_sum = new_array[R2+1] - new_array[L2]

                if left_sum == right_sum:
                    return i

        # if no values are fulfilled
        return -1
```

```python
s = Solution()

nums = [1,7,3,6,5,6]
# Output: 3

nums = [1,2,3]
# Output: -1

nums = [2,1,-1]
# Output: 0

nums = [-1,-1,0,1,1,0]
# Output: 5

s.pivotIndex(nums)
```

```python
class Solution(object):
    def pivotIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # initialise
        leftSum, rightSum = 0, sum(nums)

        # traverse through the elemtns in the loop
        for index, elemetns in enumerate(nums):
            
            # the rightsum decreases as the index moves to the right
            rightSum -= elemetns

            # it does not add the leftsum yet as the left sum is exclusive to the current index
            if leftSum == rightSum:
                return index
            
            leftSum += elemetns
    
        # if all the funcctions fail, it means it was unable to find an index that satisifeis the condtiions
        return -1
```

```python
s = Solution()

nums = [1,7,3,6,5,6]
# Output: 3

nums = [1,2,3]
# Output: -1

nums = [2,1,-1]
# Output: 0

nums = [-1,-1,0,1,1,0]
# Output: 5

s.pivotIndex(nums)
```
