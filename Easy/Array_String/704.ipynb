```python
from typing import List

class Solution:
    def search(self, nums: List[int], target: int) -> int:

        if target in nums:
            # using index function to find
            return nums.index(target)
        
        # it does not exist
        else:
            return -1
        
# complexity of O(n) from target in nums
# and O(n) in using index
# therefore final answer is O(n+n) = O(2n) which just rounds to O(n)
```

```python
sol = Solution()
# Input: 
nums = [-1,0,3,5,9,12]
target = 9

sol.search(nums,target)
# Output: 4
# Explanation: 9 exists in nums and its index is 4
```

```python
from typing import List

# ITERATIVE BINARY SEARCH
class Solution:
    def search(self, nums: List[int], target: int) -> int:

        # it is already sorted in ascending

        left = 0
        right = len(nums)

        while right > left:

            # // means round to nearest int
            middle = left + ((right-left)//2)

            # if the middle value is bigger than the target means gotta search left side (the middle value OVERSHOT LERH)
            if nums[middle] > target:

                # shift the right boundary to the middle
                right = middle - 1

            # if the middle value is smaller than the target (the middle UNDERSHOT-ed) search ride side
            elif nums[middle] < target:  

                # move the left marker to the right of the middle
                left = 1 + middle

            # target found
            else:
                # can return either left or middle
                return left
        
        return -1
```

```python
sol = Solution()
# Input: 
nums = [5, 7, 9]
target = 5

sol.search(nums,target)
# Output: 4
# Explanation: 9 exists in nums and its index is 4
```
