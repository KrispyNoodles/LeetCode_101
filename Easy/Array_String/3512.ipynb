# 3512. Minimum Operations to Make Array Sum Divisible by K

```python
from typing import List

class Solution:
    def minOperations(self, nums: List[int], k: int) -> int:

        # you can only minus to get the total_sum to be a multiple of k

        # finidng the total sum and remmiander
        total_sum = sum(nums)
        remmainder = total_sum%k

        # if total_sum is lesser than k then return the sum of the total_sum 
        # as that is the fastes you can reduce to 0
        if total_sum < k:
            return total_sum

        # however if the total_sum is already a multiple of the k,
        # then return 0
        elif remmainder == 0:
            return 0

        # lastly finding which value can be reduced to be a multiple of k
        else:
            # you will have to minus off the remmainder as operations to get a multiple of k
            # dont need to care which number to minus it from
            return remmainder
```

```python
# Input: 
nums = [3,9,7]
k = 5

sol = Solution()
sol.minOperations(nums, k)

# Output: 4
```
