# 485. Max Consecutive Ones

# Exploration

```python
from typing import List

# solution 1
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:

        counter = 0
        max_counter = 0

        for i in nums:
            if i == 1:

                counter+=1

                # if not the max_counter will reset to 1 when it sees the next 1
                # this is to ensure it is only keeping the largest counters in the row
                if max_counter <= counter:
                    max_counter = counter

            # resetting the counter the momment it sees a 0
            else:
                counter=0
        
        return max_counter
```

```python
s = Solution()
nums = [1,0,1,1,0,1]

s.findMaxConsecutiveOnes(nums)

# Output: 2
```

```python
# moving the range like sliding window across the array
# SO SLOW LOL

from typing import List

# solution 1
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:

        max_counter = 0
        left = 0

        # 0, 1, 2, 3, .....
        for right in range(len(nums)):

            # moving the window by right by 1 if the value is 1
            if nums[right] == 1:

                # take the largest value between the current incremental count and the previous max count
                max_counter = max(max_counter, right-left+1)
            
            else:
                # if the current value of right is 0, hope that the value of the next one is 1 and that is where the left index would start
                left = right + 1
        
        # this will be the value of 
        return (max_counter)
```
