```python
from typing import List

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:

        sum_value = 0

        # covnert
        for index in range(1, len(digits)+1):

            # add the 0's in as a str
            actual_value = str(digits[-index]) + ((index-1) * '0')

            # add as a value
            sum_value+=int(actual_value)

        # add 1
        sum_value+=1
        answer_array = []

        # for all charcter in the string
        for character in str(sum_value):

            # append the answer as a 
            answer_array.append(int(character))

        return answer_array
    
# time complexity of O(n) but dangerous since it is converting to many data types too many times
```

```python
from typing import List

# other people solution
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:

        # decreasing
        # the len-1 since the len is larger than the index it can shift because of 0-indexing
        for index in range(len(digits)-1,-1,-1):

            value = digits[index]

            # if the value +1 is not 10 there is no shift required
            if value+1 != 10:
                digits[index] +=1
                return digits
            
            # the 0 keeps shifting till it finds a value that is not 10
            # else it just gets added to the beginning
            digits[index] = 0

        return [1] + digits            
```

```python
digits = [1,9]

sol = Solution()
sol.plusOne(digits)

# Output: [1,0]
# Explanation: The array represents the integer 9.
# Incrementing by one gives 9 + 1 = 10.
# Thus, the result should be [1,0].
```
