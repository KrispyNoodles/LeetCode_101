# 27. Remove Element

```python
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        
        count = 0

        index = 0

        # for all values in the array
        while index < len(nums):

            # if the array is the same as the value
            if nums[index] == val:

                # pop it then append the underscore at the end
                nums.pop(index)

                # shift the index -1 because the current index has just been removed and all elements have
                # been shifted to the left by 1
                index = index - 1
                nums.append('_')

                # keep track how mant values have been removed to know how manty values are left
                count+=1

            index += 1

        print(nums)

        return len(nums)-count

# time complexity of O(n^2) because of while loop and a situation where everyone all get removed and the index gets shifted by 1
# space complecity of O(1) because no new arrays were created
```

```python
# trying again
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        
        temp_arr = []

        for value in nums:
            if value != val:
                temp_arr.append(value)

        # shortening the array to fit in
        nums[:len(temp_arr)] = temp_arr

        return len(temp_arr)

# time complecity of O(n) because of for loop to append temp arr only
# space complecity of O(n)
```
