# 58. Length of Last Word

```python
# split
s = "Hello World"

array = s.split(" ")

print(array)
```

```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:

        array_word = s.split(" ")

        i = -1

        # skip all the blanks
        while True:

            if array_word[i] != "":
                return len(array_word[i])

            else:
                i -= 1

# time complexity of O(n)
```

```python
# without using split function
class Solution:
    def lengthOfLastWord(self, s: str) -> int:

        counter = 0
        counted = False

        for i in range(len(s)-1, -1,-1):

            character = s[i]

            # if it is a blank skip it
            if character == " " and counted == False:
                continue

            elif character == " " and counted == True:
                return counter

            # else start counting
            else:
                counter +=1
                counted = True

        # after evertyghin still no return
        if counter == 0:
            return len(s)
        else:
            return counter

# time complexity of O(n)
```

```python
# Input: 
s = " a"

# Output: 1
# Explanation: The last word is "a" with length 1.

sol = Solution()
sol.lengthOfLastWord(s)
```
