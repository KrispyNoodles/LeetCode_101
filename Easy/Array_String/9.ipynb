```python
# my solution
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """

        # converting from int to string
        x = str(x)

        # reversing the string
        reverse_string = x[::-1]

        # if the reverse is the same as the current then it is True
        if reverse_string == x:
            return True
        
        else:
            return False
        
```

```python
s = Solution()

x = -121
# Output: false

x = 121
# Output: True

s.isPalindrome(x)
```
