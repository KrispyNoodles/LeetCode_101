```python
class Solution:

    # return the first occurence
    def strStr(self, haystack: str, needle: str) -> int:

        index = 0 

        while index < len(haystack):

            # checkinf if the first char match
            if haystack[index] == needle[0]:

                # then check if the word matches
                if haystack[index:index+len(needle)] == needle:
                    return index

                else:
                    index+=1

            # or else just continue as usual
            else:
                index += 1

        # if after evertyhing nothing found, then reutn -1
        return -1
    
# time complexity of O(n*m) where n is the length of the haystack and m is the needle
```

```python
# using index function
class Solution:

    # return the first occurence
    def strStr(self, haystack: str, needle: str) -> int:

        if needle not in haystack:
            return -1
        
        else:
            # find the index
            return haystack.index(needle)

# time complexity of O(n+m) where n is the length of the haystack and m is the needle
```

```python
# using index function
class Solution:

    # return the first occurence
    def strStr(self, haystack: str, needle: str) -> int:

        # going through each value takes up O(n)
        for index in range(len(haystack)):

            # comparison usese O(m)
            if haystack[index: index+len(needle)] == needle:
                return index
        
        # after everything if still make it true, it means the needle isnt inside
        return -1
    
# time complexity of O(nxm) where n is the length of the haystack and m is the needle
```

```python
sol = Solution()

# Input: 
haystack = "sadbutsad"
needle = "sad"

# Output: 0
# Explanation: "sad" occurs at index 0 and 6.
# The first occurrence is at index 0, so we return 0.

sol.strStr(haystack, needle)
```
