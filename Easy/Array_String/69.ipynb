# 69. Sqrt(x)

```python
# bin seach recap
def bin_search(array, target):

    left = 0
    right = len(array)-1
    middle = (right+left)//2

    while right>=left:

        # the value is on the left side
        if array[middle] > target:
            right = middle-1

        # else the value is on the right
        else:
            left = middle+1

        middle = (right+left)//2
    
    return middle


array = [1,2,3,4,5]
target = 5

bin_search(array, target)
```

```python
# brute force
class Solution:
    def mySqrt(self, x: int) -> int:

        # 0 <= x <= 2^31 - 1
        for val in range(0,2**30):

            square_val = val*val

            next_square = (val+1)*(val+1)

            # if the value is more than the current square, return the last value
            if x > square_val and x < next_square:
                return val
            
            if x==square_val:        
                return val
            
# time complexity of O(sqrt(n)) because it calculates the square of each value before it finds the number
```

```python
# brute force
class Solution:
    def mySqrt(self, x: int) -> int:

        # to handle for values leser than 3
        if x < 2:
            return x

        left = 0
        right = x//2 # since the largest number will never be half of the x
        answer = 0

        while right>=left:
            middle = (right+left)//2
            square = middle*middle

            if square == x:
                return middle

            # the value is on the left side
            if square > x:
                right = middle-1
                

            # else the value is on the right
            else:
                left = middle+1

                # because we round down so we take the value here
                answer = middle

        return answer
```

```python
# Input: 
x = 2147395599

sol = Solution()
sol.mySqrt(x)

# Output: 46339
```

```python
# Input: 
x = 8

sol = Solution()
sol.mySqrt(x)

# Output: 2
# Explanation: The square root of 8 is 2.82842..., and since we round it down to the nearest integer, 2 is returned.
```
