```python
# comparing using a map
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:

        # creating a dict for each and checking if they are the same
        dict_s = {}
        dict_t = {}
        
        for value in s:
            # checking if it exists
            if value in dict_s:
                # if does increment it
                dict_s[value] +=1
            else:
                # initalise to be value of 1
                dict_s[value] = 1

        # repeat for t
        for value in t:
            # checking if it exists
            if value in dict_t:
                # if does increment it
                dict_t[value] +=1
            else:
                # initalise to be value of 1
                dict_t[value] = 1

        # checking for values and length to be the same
        # this doenst work as 'aaab' is not the same as 'aabb'
        if dict_s == dict_t:
            return True

        else:
            return False

# where n is the first string and m is the second
# Time complexity of algo takes O(n+m) since the larger it gets the more it has to loop through to reassign the dict
# comparison takes up O(n) too at the last part
```

```python
sol = Solution()

# Input: 
s = "anagram"
t = "nagaram"

sol.isAnagram(s,t)

# Output: true
```

```python
# sorting
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:

        # if length different reutn False, it defo wont match
        if len(s) != len(t):
            return False

        if sorted(s) == sorted(t):
            return True
        else:
            return False

# where n is the first string and m is the second
# Time complexity of algo takes O(nlog(n)+mlog(m)) since it needs to sort everything by going through all the elemtns

# sorting is a nlog(n) operation as it requires to compare and place elements rather than looking through it
```

```python
# alternative method of Hash Map

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:

        # if length different reutn False, it defo wont match
        if len(s) != len(t):
            return False
        
        # creating a dict for each and checking if they are the same
        dict_s = {}
        dict_t = {}

        # it does the check infront so it can just do the assigning here
        for index in range(len(s)):

            # creating a map for each value, resort to 0 if there is none found
            dict_s[s[index]] = 1 + dict_s.get(s[index], 0)
            dict_t[t[index]] = 1 + dict_t.get(t[index], 0)

        # checking for values and length to be the same
        # this doenst work as 'aaab' is not the same as 'aabb'
        if dict_s == dict_t:
            return True

        else:
            return False
        
# where n is the first string and m is the second
# Time complexity of algo takes O(n+m) since it needs to sort everything by going through all the elemtns

# a much more efficinet way to store
```

```python
sol = Solution()

# Input: 
s = "anagram"
t = "nagaram"

sol.isAnagram(s,t)

# Output: true
```
