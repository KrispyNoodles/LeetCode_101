# 14. Longest Common Prefix

```python
strs = ["ab","a"]

# finding mininum
min(strs)
```

```python
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:

        # use the shortest word as example
        short_word = min(strs)

        i = 0

        for character in short_word:
            for word in strs:

                # if the chracter is not equal to the word
                if character != word[i]:
                    if i == 0:
                        return ""

                    # return i since the index is 0
                    else:
                        return word[:i]
            # icnrement i
            i += 1
        
        # if pass evertyhing
        return short_word

    # time complexity of O(nxm) it loops through all the words in the list for each charcter in the word
    # space complexity of O(m) where m is the length of the shortest word or O(1) since it is quite insignificatn
```

```python
sol = Solution()
# Input: 
strs = ["aaa","aa","aaa"]

sol.longestCommonPrefix(strs)
# Output: "aa"
```

```python
sol = Solution()
# Input: 
strs = ["ab","a"]

sol.longestCommonPrefix(strs)
# Output: "a"
```

```python
sol = Solution()
# Input: 
strs = ["flower","flow","flight"]

sol.longestCommonPrefix(strs)
# Output: "fl"
```

```python

# Input: 
strs = ["dog","racecar","car"]

sol.longestCommonPrefix(strs)
# Output: ""
# Explanation: There is no common prefix among the input strings.
```
