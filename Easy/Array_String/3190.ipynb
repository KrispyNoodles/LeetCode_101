# 3190. Find Minimum Operations to Make All Elements Divisible by Three


# Exploration

```python
# can either add 1 or -1

# remminder is 2 then (+1) is needed to get a whole number without remmainder
val = 17

print(val%3)

# if the remminder is 1 then (-1) is needed to get a whole number without remmainder
val = 10

print(val%3)
```

```python
from typing import List

# my soln
class Solution:
    def minimumOperations(self, nums: List[int]) -> int:
        
        operations = 0

        # checking if the elemtn is divisible by 3:
        for value in nums:

            if value%3 == 2:
                operations+=1

            # the opereation is 1 away from being a 
            elif value%3 == 1:
                operations+=1

            else:
                continue

            return operations
```

```python
s = Solution()

nums = [1,2,3,4]
s.minimumOperations(nums)
```

```python
from typing import List

# my soln (further simplified)
class Solution:
    def minimumOperations(self, nums: List[int]) -> int:
        
        operations = 0

        # checking if the elemtn is divisible by 3:
        for value in nums:
            
            # no remmainder, meaning no operation is required
            if value%3 == 0:
                continue

            # an operation is needed to be made
            else:
                operations+=1

        return operations
```

```python
s = Solution()

nums = [1,2,3,4]
s.minimumOperations(nums)
```
