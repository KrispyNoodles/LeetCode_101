# 392. Is Subsequence

```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:

        # if s is blank just return True
        if s == "":
            return True

        # reverse to append it
        s = s[::-1]
        stack = []

        # convert it into an array
        for char in s:
            stack.append(char)  
        
        for character in t:

            if character == stack[-1]:
                stack.pop()

            # early return
            if not stack:
                return True

        # checking if s is left
        return False
        
```

```python
# Input: 
s = "b"

t = "c"
# Output: false

sol = Solution()
sol.isSubsequence(s,t)
```

```python
# Input: 
s = ""

t = "ahbgdc"
# Output: True

sol = Solution()
sol.isSubsequence(s,t)
```

```python
# Input: 
s = "axc"

t = "ahbgdc"
# Output: false

sol = Solution()
sol.isSubsequence(s,t)
```

```python
# Input: 
s = "abc"
t = "ahbgdc"
# Output: true

sol = Solution()
sol.isSubsequence(s,t)
```
