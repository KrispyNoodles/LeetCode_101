# 409. Longest Palindrome

```python

from collections import defaultdict

class Solution:
    def longestPalindrome(self, s: str) -> int:

        apha_dict = defaultdict(int)

        # creating a count first of each alphabet
        for char in s:
            apha_dict[char] +=1

        answer = 0
        added_extra = True

        for value in apha_dict.values():

            # if the value is a multiple of 2 just add it in
            if value%2 == 0:
                answer+=value

            # to add bigg odd numbers
            else:
                if added_extra:
                    answer+=(value)

                    # to add a single 1 for the centre, can only be added once
                    added_extra = False

                else:
                    answer+=(value-1)

        return answer

# time complexity of O(n) having to create a dict
```
