# 231. Power of Two

```python
n = 0
n%2
```

```python
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        
        # all negative and 0 can eat shit
        if n<=0:
            return False
        
        # if still can divide by 2 then continue to vall function
        if n%2 == 0:
            answer = self.isPowerOfTwo(n//2)
            return answer

        else:
            if n<2:
                return True
            else:
                return False
```

```python
# Input: 
n = 0

# Output: true
# Explanation: 20 = 1

sol = Solution()
sol.isPowerOfTwo(n)
```

```python
# Input: 
n = 6

# Output: False

sol = Solution()
sol.isPowerOfTwo(n)
```
