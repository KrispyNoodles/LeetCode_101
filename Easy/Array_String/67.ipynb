```python
# my solution
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """

        value_a = 0
        value_b = 0
        
        # flip string so incremenetal is correct
        a = a[::-1]
        b = b[::-1]

        # find the value of a
        for index, value in enumerate(a):
            if value == '1':
                value_a += pow(2,index)                
        
        # find the value of b
        for index, value in enumerate(b):
            if value == '1':
                value_b += pow(2,index)
        
        # converting into binary
        bin_val = bin(value_a + value_b)

        # retriving without the '0b' in the front
        return bin_val[2:]
```

```python
s = Solution()

a = "1010"
b = "1011"
# Output: "10101"

a = "11"
b = "1"
# Output: "100"

s.addBinary(a,b)
```

```python
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """
        
        # converting into int for sum
        a = int(a, 2)
        b = int(b, 2)
        
        # converting into binary
        bin_val = bin(a+b)

        # retriving without the '0b' in the front
        return bin_val[2:]
```

```python
s = Solution()

a = "1010"
b = "1011"
# Output: "10101"

# a = "11"
# b = "1"
# Output: "100"

s.addBinary(a,b)
```
