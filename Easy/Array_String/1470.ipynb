# 1470. Shuffle the Array

### Problem  
You are given an array `nums` containing `2n` elements in the form:  
`[x1, x2, ..., xn, y1, y2, ..., yn]`.

Return the array in the new order:  
`[x1, y1, x2, y2, ..., xn, yn]`.

---

### Example 1
**Input:** `nums = [2,5,1,3,4,7], n = 3` **Output:** `[2,3,5,4,1,7]`

**Explanation:**  
`x = [2,5,1]`  
`y = [3,4,7]`  
Merged â†’ `[2,3,5,4,1,7]`

---

### Example 2
**Input:** `nums = [1,2,3,4,4,3,2,1], n = 4` **Output:** `[1,4,2,3,3,2,4,1]`

---

### Example 3
**Input:** `nums = [1,1,2,2], n = 2` **Output:** `[1,2,1,2]`

---

### Constraints
- `1 <= n <= 500`  
- `nums.length == 2n`  
- `1 <= nums[i] <= 10^3`


# Exploration

```python
nums = [1,2,3,4,4,3,2,1]

# recapping range/len to see if it starts from 0 or 1
for i in range(len(nums)):
    print(i)
```

```python
from typing import List

# Solution 1
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        # Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].
        # Return [x1,y1,x2,y2,...,xn,yn]

        front_list = nums[0:n]
        back_list = nums[n:len(nums)]

        final_list = []

        for i in range(n):
            final_list.append(front_list[i])
            final_list.append(back_list[i])

        return final_list
```

```python
# nums array is always double of n
nums = [1,2,3,4,4,3,2,1]
n = 4

s = Solution()
s.shuffle(nums, n)

# Output: [1,4,2,3,3,2,4,1]
```

```python
from typing import List

# Solution 2
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        # Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].
        # Return [x1,y1,x2,y2,...,xn,yn]

        final_list = []

        for i in range(n):
            
            # there is no need to create a duplicate array
            final_list.append(nums[i])
            final_list.append(nums[i+n])
            
        return final_list
```

```python
# nums array is always double of n
nums = [1,2,3,4,4,3,2,1]
n = 4

s = Solution()
s.shuffle(nums, n)

# Output: [1,4,2,3,3,2,4,1]
```
