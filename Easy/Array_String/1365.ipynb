# 1365. How Many Numbers Are Smaller Than the Current Number

### Problem  
You are given an array `nums`.  
For each element `nums[i]`, count how many numbers in the array are **strictly smaller** than it.  
Formally, count how many indices `j` satisfy:

- `j != i`  
- `nums[j] < nums[i]`

Return the result as an array.

---

### Example 1  
**Input:** `nums = [8,1,2,2,3]` **Output:** `[4,0,1,1,3]`

**Explanation:**  
- 8 → smaller numbers are 1, 2, 2, 3 → 4  
- 1 → none → 0  
- 2 → only 1 → 1  
- 2 → only 1 → 1  
- 3 → 1, 2, 2 → 3  

---

### Example 2  
**Input:** `nums = [6,5,4,8]` **Output:** `[2,1,0,3]`

---

### Example 3  
**Input:** `nums = [7,7,7,7]` **Output:** `[0,0,0,0]`

---

### Constraints  
- `2 <= nums.length <= 500`  
- `0 <= nums[i] <= 100`


# Exploration

```python
from typing import List

class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:

        final_array = []

        for i in range(len(nums)):
            
            # creating a variable to store the counter
            counter = 0

            # creating a temp array without the original value
            temp_array = nums[:]

            # removing the current position's value
            temp_array.pop(i)

            for val in temp_array:

                # if the current value is larger than the value being looped 
                # in the temp_array, increment it
                if nums[i]> val:
                    counter+=1
            
            # appedning all the counters
            final_array.append(counter)
        
        return final_array
```

```python
s = Solution()

nums = [8,1,2,2,3]
nums[4]
s.smallerNumbersThanCurrent(nums)

# Output: [4,0,1,1,3]
```
