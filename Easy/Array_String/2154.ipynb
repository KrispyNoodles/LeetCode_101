# 2154. Keep Multiplying Found Values by Two


# Exploration

```python
from typing import List

# solution 1
class Solution:
    def findFinalValue(self, nums: List[int], original: int) -> int:

        # looping through as many times as the values are avail in the array
        for _ in range(len(nums)):

            # checking if the current/new original value is in the array
            if original in nums:
                original = original * 2

            # skipping if the new duplicated value is not found in it
            else:
                continue
            

        return original
```

```python
s = Solution()

nums = [5,3,6,1,12]
original = 3

s.findFinalValue(nums, original)
# output:4
```

```python
from typing import List

# solution 2
class Solution:
    def findFinalValue(self, nums: List[int], original: int) -> int:

        # creating an infinite loop
        while True:
            
            # if the original is found in the array
            if original in nums:

                # the new original becomes the old one * 2
                original = original * 2

            else:
                # if the original isnt found anymore in the arrya, the previous value is the final value
                break

        return original
```

```python
s = Solution()

nums = [5,3,6,1,12]
original = 3

s.findFinalValue(nums, original)
# output: 24
```

```python
s = Solution()

nums = [5,3,6,1,12]
original = 3

s.findFinalValue(nums, original)
# output: 24
```

```python
# creating a hash table

from typing import List

# Hash table
class Solution:
    def findFinalValue(self, nums: List[int], original: int) -> int:

        # converting the list from an array into a set
        s = set(nums)

        while original in s:
            original = original * 2
            
        return original
    
# where as when a set is created, it jumps direclty into the bucket where the value should be
```
