# 3074. Apple Redistribution into Boxes

```python
from typing import List

class Solution:
    def minimumBoxes(self, apple: List[int], capacity: List[int]) -> int:
        
        # add all the apples together
        total_sum = sum(apple)

        # sort capcity from big to small then keep minusing till can
        capacity.sort(reverse=False)

        # counter to track how many boxes used
        counter = 0

        # while total_sum not negative and still crates in capcity
        while total_sum > 0 and capacity:

            # remove the last value
            pop_val = capacity.pop()

            # decrease the total sum
            total_sum = total_sum - pop_val

            # return how many times this process has been done
            counter+=1

        return counter
    
# time complecity of O(nlog(n)) because of sorted array and going through once
# space complxity of O(1) because no new list was created and I am just working wiht the current list
```

```python
apple = [9,8,8,2,3,1,6]
capacity = [10,1,4,10,8,5]
sol = Solution()
sol.minimumBoxes(apple, capacity)

# output: 5
```

```python
apple = [1,3,2]
capacity = [4,3,1,5,2]

sol = Solution()
sol.minimumBoxes(apple, capacity)

# output: 2
```
