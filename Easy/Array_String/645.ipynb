# 645. Set Mismatch

### Problem  
You originally have a set `s` containing numbers from `1` to `n`.  
Due to an error, **one number is duplicated** in the array and **one number is missing**.

You are given `nums`, the array representing the corrupted set.  
Find:

- the **number that occurs twice**  
- the **number that is missing**

Return them as an array.

---

### Example 1
**Input:** `nums = [1,2,2,4]` **Output:** `[2,3]`

---

### Example 2
**Input:** `nums = [1,1]` **Output:** `[1,2]`

---

### Constraints
- `2 <= nums.length <= 10^4`  
- `1 <= nums[i] <= 10^4`

```python
from typing import List

# the missing value is not necessarily sequenetial wth LOL poor instructions sia
# Solution 1
class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:

        # sort the list ascendingly
        nums.sort()

        length = len(nums)

        for index in range(len(nums)):
            
            # finding the duplicated value (there is only one)
            if nums[index] == nums[index-1]:
                dup_val = nums[index]
                print(f"dupe value is {dup_val}")
                break

        # now find the missing val
        # perfect set
        perf_set = set(range(1, len(nums)+1))

        # missing val (finding the difference)
        missing_val = perf_set - set(nums)

        # converting the set into a str
        val = "".join(str(x) for x in missing_val)

        return [dup_val, int(val)]
```

```python
s = Solution()

nums = [1,5,3,2,2,7,6,4,8,9]

# Output: [2, 10]

s.findErrorNums(nums)
```

```python
from typing import List

# the missing value is not necessarily sequenetial wth LOL poor instructions sia
# Solution 2
# However the code is clean but too inefficient

class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:

        # now find the missing val
        # perfect set (from 1 to the len+1 because there is a duplicate)
        perf_set = set(range(1, len(nums)+1))

        # # missing val (finding the difference)
        # # (perf set) - (set of current list)
        # missing_val = perf_set - set(nums)

        # # converting the set into a str
        # val = missing_val.pop()
        
        # alteratively (adding the perf) - (adding the unique of current list)
        list_sum = sum(set(nums))
        val = sum(perf_set) - list_sum

        # finding the dup val by calc the difference
        # beteween the set that contains unique and the original array that contains duplicates
        dup_val = sum(nums) - list_sum

        return [dup_val, val]
```

```python
from typing import List

# Approach 2 (Brute Force)
# super slow LOL but it makes sense

class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:
        
        # initialisiing variables
        dup, missing = -1, -1

        # from 0 to the highest value of the correct list
        for i in range(0, len(nums)+1):

            # counting the number of times it appear in the array
            count = nums.count(i)

            # if appears twice, it means it is the duplicated value
            if count == 2:
                dup = i

            # if appears none at all, it means it is the value that is missing
            if count == 0:
                missing = i

        return [dup, missing]        
```

```python
s = Solution()

nums = [1,5,3,2,2,7,6,4,8,9]

# Output: [2, 10]

s.findErrorNums(nums)
```

```python
from typing import List

# Approach 2 (Vector)

class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:

        # initialisiing variables
        dup, missing = -1, -1


        # creating an array of 0s the length of an ideal array
        vector = [0] * (len(nums) + 1)

        for val in nums:

            # for the values that appeared, increment that value's count by 1
            vector[val] += 1


        # the vector len will be the len of the ideal vector
        # so it can start from 1 instead of 0 since there are no 0 values in an ideal vector
        # for example [1, 2, 3, 4] no such thing as [0, 1, ....]
        for i in range(1, len(vector)):

            # duplicated found
            if vector[i] == 2:
                dup = i

            # missing found
            if vector[i] == 0:
                missing = i
        
        return [dup, missing]
```

```python
s = Solution()

nums = [1,5,3,2,2,7,6,4,8,9]

# Output: [2, 10]

s.findErrorNums(nums)
```

```python
from typing import List

# Approach 3 (map)

class Solution:
    def findErrorNums(self, nums: List[int]) -> List[int]:

        # creating a map for all values in a perfect list
        map = {i: 0 for i in range(1, len(nums)+1)}

        # initialisiing variables
        dup, missing = -1, -1

        for value in nums:
            
            # increment the value in tha map 
            map[value] +=1

        for key, value in map.items():

            # duplicated found
            if value == 2:
                dup = key

            # missing found
            if value == 0:
                missing = key

        return [dup, missing]
```

```python
s = Solution()

nums = [1,5,3,2,2,7,6,4,8,9]

# Output: [2, 10]

s.findErrorNums(nums)
```
