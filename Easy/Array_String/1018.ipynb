```python
from typing import List

# my soln
# TAKING TOO LONG TO COMPUTER TO WRONG
class Solution:
    def prefixesDivBy5(self, nums: List[int]) -> List[bool]:

        answer = []

        # current value
        for index in range(len(nums)):
            
            # retrieving from left adding incrementally
            bit_arr = nums[:index+1]

            # joinnig the array into a str
            bit_str = "".join(str(i) for i in bit_arr)

            # converting to int and doing the division
            if int(bit_str, 2)%5 == 0:
                answer.append(True)
            else:
                answer.append(False)
            
        return answer 
```

```python
from typing import List

# my soln
# using the hint
class Solution:
    def prefixesDivBy5(self, nums: List[int]) -> List[bool]:

        answer = []

        # current value
        for index in range(len(nums)):
            
            # calculating the value of the first index
            if index == 0:

                # since its either 0*2^pow(0)=0 or 1*2^pow(0)=1
                value_X = nums[0]

            else:
                # 2*value_X shifts the current total to the left by 1
                # it then adds the next number

                # the hints allows us to evaluate the final value without reassignign into an array
                value_X = 2*value_X + nums[index]

            # if the value divide by 5 has a remmainder of 0 it means it is a multiple of 5
            if value_X%5 ==0:
                answer.append(True)

            else:
                answer.append(False)
                               
        return answer 
```
