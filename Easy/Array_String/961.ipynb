# 961. N-Repeated Element in Size 2N Array

```python
from collections import defaultdict
from typing import List

class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:


        val_dict = defaultdict(int)

        for values in nums:
            val_dict[values] += 1

        # sorting based off the number of appearnce
        val_dict = sorted(val_dict.items(), key=lambda x: x[1])

        # return the largest value
        return val_dict[-1][0]

# time complexity of O(klog(k)) because of sorting where k is the number of unique elements 
```

```python
from collections import defaultdict
from typing import List

class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:

        val_dict = defaultdict(int)

        for values in nums:
            val_dict[values] += 1

            # because onyl one values appears more than once
            if val_dict[values] > 1:
                return values
            
# time complexity of O(n) because of iterating through all values in nums
```

```python
# solving using sets

from typing import List

class Solution:
    def repeatedNTimes(self, nums: List[int]) -> int:

        default_set = set()

        for value in nums:
            if value in default_set:
                return value
            
            else:
                default_set.add(value)
                
# time complexity of O(n) because of iterating through all values in nums
```

```python
# Input: 
nums = [1,2,3,3]
# Output: 3

sol = Solution()
sol.repeatedNTimes(nums)
```
