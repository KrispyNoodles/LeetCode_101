# 35. Search Insert Position

```python
# binary search recap
def bin_search(array_1, target):

    l = 0 
    r = len(array_1)-1
    middle = (r+l)//2

    while r >= l:

        # this means the target is from 0 to the middle
        if array_1[middle] > target:
            # change the bound
            r = middle-1
    
        else:
            l = middle+1

        # it has to be right + left //2 to get the correct index
        middle = (r+l)//2
    
    return middle

array_1  = [1,3,4,5,6]
target = 1
bin_search(array_1, target)
```

```python
# proper method
from typing import List

class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:

        left = 0

        # must be -1 so that the index is within
        right = len(nums)-1
        
        # this is the difference between
        while right>=left:

            # setting middle
            middle = (right+left)//2

            if nums[middle] == target:
                return middle

            # if the middle is more than the target, it is on the left side
            if nums[middle]>target:
                right = middle-1
            
            else:
                left = middle+1
        
        # return left if not found
        # it acts as the lower bound of where this new target should be placed at

        # right is the upper bound of this new target
        return left
        
# time complexity of Bin Search is O(log(n))
```
