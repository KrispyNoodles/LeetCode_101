# 206. Reverse Linked List

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# creation of Nodes
ListNode1 = ListNode(1, None)
ListNode2 = ListNode(2, ListNode1)
ListNode3 = ListNode(3, ListNode2)
ListNode4 = ListNode(4, ListNode3)
ListNode5 = ListNode(5, ListNode4)
```

```python
# OH MY GOSH I DID IT, after 3 hours

from typing import Optional

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:

        # initalising
        curr = head
        temp_1 = None

        # cant use while temp_2 != None as it assumes there are no empty head

        # as long as the head next is not none move the head
        while curr != None: 

            # commenting out print statement to run faster
            # print(f"adjusting the current node of {curr.val}")

            # retrieving for temp_2 before assining
            # this contains the next object of what is being edited
            temp_2 = curr.next

            # this assigns the current edited' object to the reversed value
            curr.next = temp_1

            # it updates itslef to be the next object to be pointed to in the next object
            temp_1 = curr

            # it shifts the current to be edited to the one that was pointed too previously
            curr = temp_2
        
        # the temp_1 will contain the new head at the end
        return temp_1
            
    def view_link_list(self, head: Optional[ListNode]) -> Optional[ListNode]:

        print(head.val)

        curr = head

        while curr.next != None:

            curr = curr.next
            print(curr.val)


# time complexity of O(n)
# space complexity of O(1) beacuse the size of the array does not affect how much extra space is needed to make each operation, where only 3 pointers are used
```
