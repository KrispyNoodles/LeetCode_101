````python
import pandas as pd

def rising_temperature(weather: pd.DataFrame) -> pd.DataFrame:

    # create a date time column called yesterday
    weather['recordDate'] = pd.to_datetime(weather['recordDate'])
    weather['yesterday'] = weather['recordDate'] - pd.Timedelta(days=-1)

    # create an inner join
    weather = weather.merge(weather, how='inner', left_on='recordDate', right_on='yesterday')

    # filter
    filter_df = weather['temperature_x']>weather['temperature_y']
    
    ans_df = weather[filter_df]

    # renaming column
    ans_df = ans_df.rename(columns={'id_x':'id'})

    return ans_df[['id']]
````
