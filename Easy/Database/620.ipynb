```python
import pandas as pd

data = [[1, 'War', 'great 3D', 8.9], [2, 'Science', 'fiction', 8.5], [3, 'irish', 'boring', 6.2], [4, 'Ice song', 'Fantacy', 8.6], [5, 'House card', 'Interesting', 9.1]]
cinema = pd.DataFrame(data, columns=['id', 'movie', 'description', 'rating']).astype({'id':'Int64', 'movie':'object', 'description':'object', 'rating':'Float64'})
```

```python
# how to sort
cinema = cinema[cinema["id"]%2!=0]
cinema = cinema.sort_values(by='id', ascending=False)
cinema
```

```python
import pandas as pd

def not_boring_movies(cinema: pd.DataFrame) -> pd.DataFrame:
    # retrieving only odd
    cinema= cinema[cinema["id"]%2!=0]

    # filtering out values with boring in the description
    cinema = cinema[cinema["description"]!='boring']

    # sorting the values descending by the rating
    cinema = cinema.sort_values(by='rating', ascending=False)

    return cinema
```

```python
not_boring_movies(cinema)
```

```python
import pandas as pd

def not_boring_movies(cinema: pd.DataFrame) -> pd.DataFrame:
    
    # brackets are needed for each condition
    cinema = cinema[(cinema['description']!='boring') & (cinema['id']%2 != 0)]

    # REMEMBER IT IS A SORT_VALUES
    cinema = cinema.sort_values('rating', ascending=False)
    return cinema
```
