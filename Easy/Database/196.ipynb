# 196. Delete Duplicate Emails

````python
import pandas as pd

data = [[1, 'john@example.com'], [2, 'bob@example.com'], [3, 'john@example.com']]
person = pd.DataFrame(data, columns=['id', 'email']).astype({'id':'int64', 'email':'object'})
````

```sql

-- this does not delete anything, therefore it is wrong
SELECT DISTINCT person.email
FROM person

```sql
-- this works where it deletes all person's id whose is not the minimum 
-- after we do a group by person.email to remove duplicates

DELETE FROM PERSON
WHERE PERSON.id NOT IN (
    SELECT MIN(id) FROM PERSON GROUP BY PERSON.email)

````python
import pandas as pd

def delete_duplicate_emails(person: pd.DataFrame) -> None:
    
    # drop duplicates subset=['email']

    # when inplace is set true, it means the original dataframe is being edited and that there is nothing returned
    person.drop_duplicates(subset=['email'], inplace=True)
    
````

````python
delete_duplicate_emails(person)
````
