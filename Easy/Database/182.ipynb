SELECT p.email
FROM Person p
# GROUP BY is to group things together based off a certain key
GROUP BY p.email
# HAVING ONLY WORKS TOGETHER AFTER A GROUP BY, WHERE CANT BE USED!!!
HAVING COUNT(p.email) > 1

```python
import pandas as pd

data = [[1, 'a@b.com'], [2, 'c@d.com'], [3, 'a@b.com']]
person = pd.DataFrame(data, columns=['id', 'email']).astype({'id':'Int64', 'email':'object'})
```

```python
# return as a index

counts = person.groupby('email').size()

# without index
# it removes the count labels that was previously created
repeated_emails = counts[counts>1]
print(repeated_emails)

# with index
repeated_emails = counts[counts>1].index
print(repeated_emails)
```

```python
def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:

    # group by the 'email' name
    # and count the number for each category
    counts = person.groupby('email').size()

    # usign the counts to remove counts<1
    repeated_emails = counts[counts>1].index

    # recreate the datafram again
    df = pd.DataFrame({"email":repeated_emails})

    return df
```

```python
duplicate_emails(person)
```

```python
# size vs count
filter = person.groupby('email').count()
print(type(filter))

filter = person.groupby('email').size()
print(type(filter))
```

```python
# redo
import pandas as pd

def duplicate_emails(person: pd.DataFrame) -> pd.DataFrame:

    # count returns as a df
    # while size reutrns as a pd series
    # size is easier to work with
    filter = person.groupby('email').size()

    # retrieves the values without the count
    counts = filter[filter>1].index

    # create a datafrmae
    df = pd.DataFrame({'email': counts})

    # return counts.index.to_frame(name='email')
    return df
```

```python
duplicate_emails(person)
```
