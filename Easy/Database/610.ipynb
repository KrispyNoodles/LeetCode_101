```sql
-- Triangle line is: Lenth of one line is less than sum of lenth 2 others lines.
SELECT *,
CASE
    WHEN x < (y+z) AND y < (x+z) AND z < (x+y) THEN 'Yes'
    ELSE 'No'
END as triangle 
FROM Triangle

````python
import pandas as pd

def triangle_judgement(triangle: pd.DataFrame) -> pd.DataFrame:

    # creating a new columns
    # axis is to say to perform that function for each row
    triangle['triangle'] = triangle.apply(is_valid, axis=1)
    return triangle

# creating a function to check for vlaid triangle
def is_valid(row):

    x = row['x']
    y = row['y']
    z = row['z']

    if x < (y+z) and y < (x+z) and z < (x+y):
        return "Yes"
    else:
        return "No"
````

````python
# rnew approach using where
import pandas as pd
import numpy as np

def triangle_judgement(triangle: pd.DataFrame) -> pd.DataFrame:
    
    # the syntax for where is 
    # np.where(condition, value_if_true, value_if_false)
    triangle['triangle'] = np.where(
                                    # the brackets are needed since & operator are higher than >
                                    # therefore it will use the & first and doing the oepration wrongly
                                    (triangle['x']+triangle['y']>triangle['z']) &
                                    (triangle['x']+triangle['z']>triangle['y']) &
                                    (triangle['z']+triangle['y']>triangle['x']),
                                    'Yes',
                                    'No'
                                    )

    return triangle
````
