# Insert Constraints
# if it violates the ENUM such as it needing to be one of the 3 strings 
# or if it does not have a name
# or if no id
# or if the age is younger than 18

# an error will appear such as
null value in column "id" of relation "users" violates not-null constraint

CREATE TYPE status AS ENUM ('active', 'inactive', 'pending');

CREATE TABLE users (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER CHECK (age >= 18),
    status status
);
