# 619. Biggest Single Number

```sql
-- Write your PostgreSQL query statement below

SELECT MAX(num) AS num 
FROM (
    SELECT num 
    FROM MyNumbers
    GROUP BY num
    HAVING COUNT(num) = 1
) AS subquery;

````python
import pandas as pd

data = [[8], [8], [3], [3], [1], [4], [5], [6]]
my_numbers = pd.DataFrame(data, columns=['num']).astype({'num':'Int64'})
````

````python
import pandas as pd

def biggest_single_number(my_numbers: pd.DataFrame) -> pd.DataFrame:

    # rename column
    
    # group by the value
    new_df = my_numbers.groupby(['num'])['num'].count().reset_index(name='count')

    new_df = pd.DataFrame(new_df)

    new_df = new_df[new_df['count']==1]

    # sort then retrieve the highest
    new_df = new_df.sort_values(by='num', ascending=False)

    # final ans

    # checking if df is empty, return None
    if new_df.empty:
        d = {'num': None}

        # needs index
        return pd.DataFrame(data=d, index=[0])

    # limit by 1 and return the num col only
    return new_df[:1][['num']]
````

````python
# using drop duplicates
import pandas as pd

def biggest_single_number(my_numbers: pd.DataFrame) -> pd.DataFrame:

    # rename column
    
    # group by the value
    new_df = my_numbers.drop_duplicates(keep=False)

    if not new_df.empty:

        answer = new_df.max().reset_index(name='num')
        return pd.DataFrame(answer)[['num']]
    
    else:
        return pd.DataFrame({'num':None}, index=[0])
````
