```python
import pandas as pd

data = [[1, 'Joe', 70000, 3], [2, 'Henry', 80000, 4], [3, 'Sam', 60000, None], [4, 'Max', 90000, None]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'managerId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'managerId':'Int64'})
```

```python
import pandas as pd

# finding which employee earn more than their boss
def find_employees(employee: pd.DataFrame) -> pd.DataFrame:

    # add in row of salary of manager
    df = pd.merge(employee, 
                          employee, 
                          left_on='managerId', 
                          right_on='id',

                          # suffix is what is added behind to know the difference
                          suffixes=('', '_manager'))
    
    # filtering employee salary > manager salary
    result = df[df['salary'] > df['salary_manager']]

    # syntax for renaming
    # changing from column from 'name' to 'Employee'
    result = result.rename(columns={'name': 'Employee'})

    # returning only the column called '
    return result[['Employee']]
```

```python
find_employees(employee)
```

```python
# REDO
import pandas as pd

def find_employees(employee: pd.DataFrame) -> pd.DataFrame:
    
    # doing a left join
    employee = employee.merge(employee, how='left', left_on='managerId', right_on='id')

    # salary x more than salary 
    employee = employee[employee['salary_x']>employee['salary_y']]

    # do a rename
    employee = employee.rename(columns={"name_x": "Employee"})

    # then do a sort
    return employee[["Employee"]]
```
