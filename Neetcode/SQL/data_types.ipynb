# ENUM
# it restricts the user to a set of predefined values similar to the restrain of "CHECK value IN(string_1, string_2)"

CREATE TYPE account_type AS ENUM ('checking', 'savings', 'cd', 'money_market');

CREATE TABLE bank_accounts (
  id INTEGER PRIMARY KEY,
  account_type account_type,
  balance INTEGER
);



# CREATING AN AUTO INCREMENT SEQUENCE 

# SERIAL is the most common method for auto-incrementing columns in Postgre SQL but might not work for others
# GENERATED ALWAYS AS IDENTITY is the SQL standard way of doing it
# SEQUENCE (WORKS SIMILARLY TO ENUM) which can be used by other columns

CREATE SEQUENCE gov_id START WITH 1000 INCREMENT BY 3;

CREATE TABLE gov_employee (
  id INTEGER GENERATED ALWAYS AS IDENTITY PRIMARY KEY ,
  gov_id INTEGER DEFAULT nextval('gov_id'),
  name TEXT
);


# GENERATING A UUID (UNIVERSALLY UNIQUE IDENTIFIER)

# command to genrate is 'uuid_generate_v4()'

CREATE TABLE users (
    id UUID PRIMARY KEY DEFAULT gen_random_uuid(),
    name TEXT
);
