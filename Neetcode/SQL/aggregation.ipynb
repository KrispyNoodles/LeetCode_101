# LENGTH can be used to count the length of the string
SELECT COUNT(DISTINCT(author)) as unique_authors
FROM books
WHERE LENGTH(author) <12;

# alterantively, this ensures that each author minimally must have 12 characters
SELECT COUNT(DISTINCT(author)) as unique_authors
FROM books
WHERE author NOT LIKE '____________%';

# AGGREGATION MIN MAX

# MIN and MAX and COUNT can be used all together in the same query

SELECT ROUND(MIN(score)) as min_score, 
ROUND(MAX(score)) as max_score, 
ROUND(AVG(score)) as avg_score
FROM scores
WHERE region='EU';

# AGGREGATIION GROUP BY

# if we have a table of books, we could get the number of books for a specific author like this
SELECT COUNT(*)
FROM books
WHERE author = 'John Doe'

# but if we want to get the number of books for every author, we can use the GROUP BY clause
SELECT author, COUNT(*)
FROM books
GROUP BY author

# it can be used togethere with other aggregation methods
SELECT country, SUM(gold) as gold_medals, 
SUM(silver) as silver_medals,
SUM(bronze) as bronze_medals
FROM olympic_medals
GROUP BY country

# AGGREGATION HAVING
# this is used together with the GROUP BY CLAUSE
# It is mainly used to filter the results of a GROUP BY clause

SELECT book_name
FROM book_sales
GROUP BY book_name
HAVING SUM(copies_sold) > 500;


# AGGREGATION EXECUTION ORDER

SELECT book_name
FROM book_sales
WHERE book_name LIKE 'A%'
GROUP BY book_name
HAVING SUM(copies_sold) > 100
ORDER BY book_name
LIMIT 1;

# based off this query

# the FROM is slected first
# then the WHERE

# then the GROUP BY
# and then the HAVING SUM

# followed by the ORDER BY
# then the LIMIT of 1 

# and finally the SELECT


SELECT country, year, SUM(gold + silver + bronze) as total_medals
FROM olympic_medals 
WHERE category != 'Gymnastics'
GROUP BY country, year
HAVING SUM(gold + silver + bronze) >20

# ORDER BY can use the alias defined in the SELECT clause
ORDER BY total_medals DESC
LIMIT 5;
