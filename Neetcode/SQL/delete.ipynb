# Deleting rows from a table

# unlike the SELECT AND UPDATE stamatemnts, the DELETE statement always operatees on all columns of a row selected to be remove
# this means we can only delete entire rows and not specific columns

# deletes all rows in the table
DELETE FROM table_name

# conditional deleting
DELETE FROM students
WHERE age <18 OR name LIKE 'A%';

# this removes all rows with age below 18 and anyone with the name starting with an A

# RETURNING CLAUSE can also be used
# this checks which rows have been deleted

DELETE FROM bank_accounts
WHERE balance<0 AND last_transaction_date < '2024-01-01'
RETURNING id, name;

# the constraints applied earlier when creating a table which affects the INSERT and UPDATE also affects the DELETE
