# the 'UNIQUE' CONSTRAINT

# this means that whenever there is an id added into the table it has to be unique

CREATE TABLE students(
  id INTEGER UNIQUE NOT NULL,
  name TEXT NOT NULL,
  age INTEGER NOT NULL
)

# PRIMARY KEY

# a PRIMARY KEY IS SPECIAL COLUMN within the table and it must be UNIQUE and cannot contain a NULL VALUE

CREATE TABLE users(
  id INTEGER PRIMARY KEY,
  username TEXT
)

# FOREIGN KEY

# a FOREIGN KEY IS IS A COLUMN that references to a column in another table

# this means then whenever a value is inserted into the owner_id of videos, that same value must be found in the column of the users id
# if not it will fail

# IT IS NOT STATED "FOREIGN KEY" but it is called "REFERENCES"

CREATE TABLE users (
    id INTEGER PRIMARY KEY,
    name TEXT
);

CREATE TABLE videos (
    id INTEGER PRIMARY KEY,
    title TEXT,
    owner_id INTEGER REFERENCES users(id)
);


# COMPOSITE PRIMARY KEY

# a COMPOSITE PRIMARY KEY is made up of different columns in the same table

# this means then whenever a value is inserted into the owner_id of videos, that same value must be found in the column of the users id
# if not it will fail

# IT IS NOT STATED "FOREIGN KEY" but it is called "REFERENCES"

CREATE TABLE orders(
  order_id INTEGER,
  product_id INTEGER,
  quantity INTEGER,
  PRIMARY KEY (order_id, product_id)
);
