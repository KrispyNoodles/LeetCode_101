```python
# creating stack from array

from typing import List

class Solution:
    def countStudents(self, students: List[int], sandwiches: List[int]) -> int:

        counter = 0

        # continue running as long as both arrays exis
        while students and sandwiches and counter <= len(students):
            
            # if counter goes through all
            counter += 1
            
            # if the first value matches
            if students[0] == sandwiches[0]:
                # remove both student and sandwich
                students = students[1:]
                sandwiches = sandwiches[1:]

                # reset the counter
                counter = 0
            
            # else move the student to the back of the list
            else:
                new_students = students[1:]
                new_students.append(students[0])

                students = new_students[:]

        if students:
            return len(students)
        else:
            return 0
```

```python
from typing import List

# creating a stack from linked list
# Definition for douvly-linked list.
class ListNode:
    def __init__(self, val=int, next=None, prev=None):
        self.val = val
        self.next = next
        self.prev = prev

class Solution:

    def __init__(self):

        # creating a head and a tail of the queue
        self.head = None
        self.tail = None

    def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
        
        # creating a stack for the students
        for student in students:
            self.add_stack(student)

        # now loop through and if it is the correct val that it will be remove from the stack
        counter = 1

        while self.head != None and sandwiches and counter <= len(students):
            
            if self.head.val == sandwiches[0]:

                # remove the node from the left
                self.pop_left()

                # adjust the sandwicha rray
                sandwiches = sandwiches[1:]

                # reset the counter
                counter = 0

            else:                
                # pop the left value and add it back to the stack
                value_pop = self.pop_left()
                self.add_stack(value_pop)

                # increment by 1
                counter += 1

        return self.counter()


    def add_stack(self, value):
        node_added = ListNode(value)

        # if it is empty
        if self.isEmpty():
            self.head = node_added
            self.tail = node_added
            return
        
        # else...
        self.tail.next = node_added

        # add the previous for the node_added back to the old tail
        # typo of using tail instead of prev
        node_added.prev = self.tail

        self.tail = node_added


    def pop_left(self) -> int:
        if self.isEmpty():
            return None

        # pop_val
        pop_left_val = self.head.val

        # if there is only one element left, reset
        if self.head.next is None:
            self.tail = None
            self.head = None
            return pop_left_val
        
        # moving the head node
        self.head = self.head.next
        self.head.prev = None

        return pop_left_val
     
     # counting how many values are left in the stack
    def counter(self) -> int:

        curr = self.head
        count = 0
        while curr != None:
            count += 1
            curr = curr.next

        return count

    def isEmpty(self) -> bool:
        
        if self.head == None:
            return True
        else:
            return False
```
