```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None, prev=None):
        self.val = val
        self.next = next
        self.prev = prev
```

```python
# Double linked Link

# this works the same just that now for each node it has a prev value

# to add a new node to the end
new_node = ListNode(-1)
tail.next = new_node
new_node.prev = tail
tail = new_node

# removing a node
# node --> node_to_be_removed
new_tail_node = tail.prev
tail = new_tail_node
tail.next = None

# the double link list can also be used as a stack because of its push and pop qualities of taking up an operation of O(1)
# however the downside is accessing an elememt, it cant be done with just an index and has to move across from the head or tail
# which will be an O(n) operation
```

## SUMMARY OF ARRAYS vs LINKED LISTS

| | Arrays | LinkedLists |
|---|---|---|
| Operation | Big-O Time | Big-O Time |
| Acess i-th Element | O(1) | O(n) |
| Insert/Remove End | O(1) | O(1) |
| Insert/Remove Middles | O(n) | O(1) |
| Insert/Remove Middles by Index | O(n) | O(n) |

The O(1) operation for inserting and removing in the middle assumes you already have the pointer to which node that is being removed
or the node that it is being added beside to

linked lists uses O(1) to remove because it does not require to shift the other elements
