```python
# Linked list are similar to arrays
# A ListNode object contains two things minially which is it's value and the pointer of the next ListNode
# the next node is what allows the ListNode to connect to the next object

# link list are not stored in the same way we used them as programmers (as in sequentially)
# they can be in some random order in memory but connected through pointers
# therefore this is the biggest difference in comparison to arrays in the sense that they are stored differently in memory
```
