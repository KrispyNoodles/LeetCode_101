```python
# what is a data structure
# it is a way of structuring data from the perspective of a computer
# the array and variables are stored into the ram (it takes in Byters, such as 8GB ram or 16GB ram)

# each integers of the arrray are converted into 4 Bytes, 32 bit Int
# for arrays we do not know which address the values will be stored in, but we know that the values within will be continous

# each ASCII characters only needs 1 Byte 
```

```python
# STATIC ARRAYS PROPERTIES
# Properties of an array
# index starts at 0
my_array = ['a', 'b', 'c']

# when this happens we are doing a RAM operation (Random Access Memory) 
# we do need to start from the beginning of the address and search for it sequentially but we can find it immediately,
# whiich is why it only takes up O(1)
my_array[0]

# if we were to add a value to the right of the array,
# it might not be neccesarily true then we can just add it into the address beside it, as there might be something in it already
# and if we were to just put another value anywhere else, it does not any longer follow the property of arrays of it being continuous

# THE BIGGEST LIMITATION OF Static Array (FIXED SIZ ARRAY) but python and javascript dont have this option, and it usese DYNAMIC ARRAYS as the default

# writing, reading and deleting to arrays is awlasy instant time of O(1)

# when adding a value in between of the list, we will have to ove all the values after to the next address, 
# therefore it is a possibility of a worse case of moving all the elemnts in the array leading to the operation to be O(n)
```

## SUMMARY OF STATIC ARRAYS 

inserting is using `append` to add to the end and `pop` to remove 

inserting/removing values in the middle of the array requires shifting the other elements to cover


| Operation | Big-O Time |
|---|---|
| Read/ Write with index | O(1) |
| Insert/Remove End | O(1) |
| Insert Middles | O(n) |
| Remove Middles | O(n) |
