# Recursion

# One-Branch recursion
n! = n * (n-1) * (n-2) .... (1)

5! = 5 * (5-1) * (5-2) .... (1)
5! = 5 * 4 * 3 * 2 * 1

# 5! = 5 * 4!


```python
# coding a factorial function

value = 5

def fact_func(value):
    
    ans = 1

    for i in range(0, 5):
        ans = ans * (value-i)

    return ans
    
fact_func(value)
```

```python
# coding a factorial function
# can also be
# 5! = 5 * 4!

# n! = n*(n-1)!

value = 5

def fact_func(value):

    # to intiate the last factorial 1!
    if value == 1:
        return 1
    
    ans = 1

    ans = value * fact_func(value-1)

    return ans
    
fact_func(value)
```

```python
# model answer
value = 5

def factorial(n):

    # this is the last term
    # which is the base case and therefore the formula of n! = n*(n-1) isnt needed
    if n == 1:
        return 1

    return n*factorial(n-1)

factorial(value)
```

```python
# reverse link list using recusrion
# Definition for singly-linked list.
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:

        if head.next == None:
            return head
        
        else:
            head.next = self.reverseList(head.next)
            return head      
```

```python
# two branch recursion
# Fibonacci Number F(n) = F(n-1) + F(n-2)
                #  F(0)=0 and F(1)=1


def fibo_numb(n):
    
    # base case
    if n<=1:
        return n
    
    return fibo_numb(n-1)+fibo_numb(n-2)
```
