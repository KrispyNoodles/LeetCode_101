# BIG O NOTATION CS
it is a method to analyse the amount of time it takes for the algo to execute as the input size grows

# O(n)
if the function computign time grows linearly with the size of the input
Then it can be just represented with O(n), where n is the input size 

if the function too grows by O(n+5) or O(n/2) we just round it to O(n)


# O(1)
This means that not matter how much our input size grow the time complexity of the Big O of one-time is alawys the same
(these are he most efficientt algos!!)

```python
# An example of these operations are... O(1)

# Because the runtime of these operations does not increase with the size of the input, they are considered O(1)!!

# append and pop of arrays
nums = [1,2,3,4]
nums.append(5)
nums.pop()

# look up of array based of index
nums[0]
nums[1]
nums[2]


# HashMap/ Set
hashMap ={}
hashMap["key"] = 10 # isnerting value
print("key" in hashMap) #lookup based for existence
print(hashMap["key"]) #lookup based of key
hashMap.pop("key") # removing 
```

```python
# now we look at the semi most garbarge examples of O(n)

# these examples do not show O(n) but they show that they have a worst case scenerio of O(n)
# as the input increases, the computing time also increases linearly

nums = [1,2,3]
sum(nums) # sum of array because it adds everything
for n in nums: # for loops
    print(n)

nums.insert(1, 100) # insert middle, because it has to go throguh the array to put the value
# if we were to put 100, into the index 1, we will have to shift everything in the array to the right by 1

nums.remove(100) # remove values in the middle
# if we were to remove a value at the front, we will have to shift the other values to cover up to the left

print(100 in nums) # searching through

# algos that belong in here are
# SLIDING WINDOW and MONOTONIC STACK

```
