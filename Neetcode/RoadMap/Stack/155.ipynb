```python
# my solution
class MinStack:

    def __init__(self):
        self.array = []

    # adding the elemtn into the array
    def push(self, val: int) -> None:
        self.array.append(val)

    # removing the top elemnt
    def pop(self) -> None:
        self.array.pop()

    # retrieving the top elemtn
    def top(self) -> int:
        return self.array[-1] 

    # returnng the smallest value in the array
    def getMin(self) -> int:
        return min(self.array)
    
# time complexity of O(n)
# this is mainly from the minimum
# the rest of the operation is an O(1) operation since the time does not affect it

# space complexity of O(n)
# as the size of the array grows, the size required grows as well
        
```

```python
# creating another stack for the minium value
# precomputing the min value
class MinStack:

    def __init__(self):
        self.array = []

        # it does not store an order from small to big
        # but rather a small version of the main stack
        self.min_stack = []

    # adding the elemtn into the array
    def push(self, val: int) -> None:
        self.array.append(val)

        # if there is no value, the current value is the smallest
        if not self.min_stack:
            self.min_stack.append(val)
        
        else:
            # the current smallest is smaller than that value
            if self.min_stack[-1] >= val:
                self.min_stack.append(val)         

            else:
                self.min_stack.append(self.min_stack[-1])     

    # removing the top elemnt
    def pop(self) -> None:
        self.array.pop()
        self.min_stack.pop()

    # retrieving the top elemtn
    def top(self) -> int:
        return self.array[-1] 

    # returnng the smallest value in the array
    def getMin(self) -> int:
        return self.min_stack[-1]
        
```
