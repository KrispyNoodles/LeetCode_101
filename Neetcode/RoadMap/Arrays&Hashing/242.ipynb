# 242. Valid Anagram

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        
        # rearrange both arrays then compare
        if sorted(s) == sorted(t):
            return True

        else:
            return False


# where m and n are the two different strings
# Time complexity of time is: O(mlog(m) + nlog(n)
# since the longer the input is, the operation will become longer linearly

# Space complexity: O(m + n)
# increasing the size does not create any further space
```

```python
sol = Solution()

# Input: 
s = "racecar"
t = "carrace"

sol.isAnagram(s,t)

# Output: true
```
