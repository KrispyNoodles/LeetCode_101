# 217. Contains Duplicate

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        
        if len(nums) > len(set(nums)):
            return True
        else:
            return False

# complexity of time computation is O(n)
# it has to take a longer time to count the length and make the comparison
# the longer the len of the array is

# complexity of Space is O(n) 
# because as array increases, a larger set variable is created

```

```python
sol = Solution()

# Input: 
nums = [1, 2, 3, 3]

sol.hasDuplicate(nums)

# Output: true
```
