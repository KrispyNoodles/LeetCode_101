# 128. Longest Consecutive Sequence

```python
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        largest_len = 0

        # sort first the entire array from small to big
        sorted_values = sorted(nums)

        # create a temp array
        temp_arr = []
        # assume the first value is the begining of the biggest lenght

        for value in sorted_values:

            # if temp_arr is empty
            if not temp_arr:
                temp_arr.append(value)

            else:
                # check if the previous value inside is continous
                if value == temp_arr[-1] + 1:
                    temp_arr.append(value)

                # if the value is the same can just continue
                elif value == temp_arr[-1]:
                    continue

                else:
                    # reassign the max_leng
                    largest_len = max(largest_len, len(temp_arr))

                    # insert the value into the temp_arr
                    temp_arr = [value]


        # if it is just continuous, we have to compare between the empty and the temp_arr
        return max(largest_len, len(temp_arr))

# time complexity of O(nlog(n)) because of the sorted array
# space complexity of O(n) as the array grows bigger the space needed for it grows as well
```

```python
sol = Solution()

nums = [9,1,4,7,3,-1,0,5,8,-1,6]

sol.longestConsecutive(nums)
# Output: 7
```

```python
# Brute force
from typing import List

class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        # converting the list into a set
        # this removes duplicated number
        number_set = set(nums)

        result = 0

        for number in nums:

            # restart the streak count to 0
            streak_count = 0

            num_to_find = number

            # if the num_to find is inside
            # then incremenet the streak count and update the num to find
            while num_to_find in number_set:

                streak_count += 1

                # updating the num_to_find by 1
                num_to_find = num_to_find + 1

            # after every streak compare to the current max
            result = max(result, streak_count)
        
        return result
    
# however this uses O(n^2) for time complexity will face TLE (time limit exceeded)
```

```python
sol = Solution()

nums = [9,1,4,7,3,-1,0,5,8,-1,6]

sol.longestConsecutive(nums)
# Output: 7
```

```python
# similar to my answer
from typing import List

# using sorting
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        # checking for empty arrays
        if not nums:
            return 0
        
        # sort the list
        nums.sort()

        result = 0
        streak = 0

        # initalise with the first eleemtn in nums
        current = nums[0]

        i = 0
        while i < len(nums):
            # if the current value to be found
            # is not the same as the current value
            # it resets
            if current != nums[i]:
                current = nums[i]
                streak = 0
            
            # if the values are the same, countinue incrementing i
            while i < len(nums) and current == nums[i]:
                i += 1
            
            # increment the streak
            streak +=1
            current +=1

            # retrieveing the largest result
            result = max(result, streak)

        return result
```

```python
sol = Solution()

nums = [100,4,200,1,3,2]

sol.longestConsecutive(nums)
# Output: 4
```

```python
# hash set
from typing import List

# using sorting
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        # converting it inot a set
        nums_set = set(nums)
        longest_streak = 0

        for values in nums_set:
            
            # if the previous value is not inside, it is the first vallue of the streak
            if (values - 1) not in nums_set:
                counter = 1

                while (values + counter) in nums_set:
                    # as long as the next value is inside, then increment to the next value
                    counter +=1
                
                # after finish counting compare to the longest_streak
                longest_streak = max(longest_streak, counter)
        
        return longest_streak
```

```python
sol = Solution()

nums = [100,4,200,1,3,2]

sol.longestConsecutive(nums)
# Output: 4
```

```python
# hash map
from typing import List
from collections import defaultdict

# using sorting
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:

        # creating a defaultdict
        map_num = defaultdict(int)
        result = 0

        for value in nums:

            # ignore duplicates to preserve the logic
            if not map_num[value]:

                # lengths of existing left and right intervals
                left_len = map_num[value - 1]
                right_len = map_num[value + 1]
                
                # total merged length including current value
                total_len = left_len + right_len + 1

                # storing the value so that it will not look into this value again
                # and it allows for future merges
                map_num[value] = total_len 

                # updating the left boundary: move left_len steps left
                map_num[value - left_len] = total_len

                # updating the right boundary: move right_len steps left
                map_num[value + right_len] = total_len

                 # updating the best result
                result = max(result,map_num[value])
        
        return result
```

```python
sol = Solution()

nums = [100,4,200,1,3,2]

sol.longestConsecutive(nums)
# Output: 4
```
