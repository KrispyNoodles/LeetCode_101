# 1. Two Sum

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
    
        index_1 = 0
        index_2 = 0

        for index in range(len(nums)): # O(n)

            # initating
            index_1 = index

            # finidng the other target 
            target_2 = target-nums[index_1]

            if target_2 in nums: # O(n)
                # find the index of target 2
                index_2 = nums.index(target_2) # O(n)

                # checking if the index is index_1 
                if index_2 != index_1:

                    # returning it as an array that is sorted
                    return sorted([index_1, index_2])


# time complecity is O(n^2)
# one from the 'for' loop
# one from checking if the value is in the array
# and the last one is from finding the index of the value in the array
# therefore it is O(n) * (O(n)+O(n)) = O(n^2) approximately

# space complexity is O(1) because the space the number of variables I allocate does not grow with more variables
```

```python
sol = Solution()

# Input: 
nums = [3,4,5,6]
target = 7

sol.twoSum(nums,target)
# Output: [0,1]
```
