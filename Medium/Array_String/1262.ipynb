```python
from typing import List

class Solution:
    def maxSumDivThree(self, nums: List[int]) -> int:

        # calculate the total number if dividable by 3

        # 3 arrays:
        # value with 0 remmainder
        # value with 1 remmainder (in increasing value)
        # value with 2 reammainder (in increasing value)
        remmainder = sum(nums)%3
        remain_1 = []
        remain_2 = []

        multiples_of_3 = []

        if remmainder==0:
            return sum(nums)
        
        else:
            for element in nums:
                if element%3 == 1:
                    remain_1.append(element)
                elif element%3 == 2:
                    remain_2.append(element)
                elif element%3 == 0:
                    multiples_of_3.append(element)
            
            # sort both arrays decreasing the increasing
            remain_1.sort()
            remain_2.sort()
            
            # this is assuming that every array will be minusing off remmainder values
            if remmainder == 1:
                
                # if there are no values in remmainder 1 or the remmainder_2 has less that 2 values in the array
                if remain_1 or len(remain_2)<2:

                    # if there are values in remainder_2 do comparison
                    if len(remain_2)>=2:
                        # either remove 1 from remain_1 or 2 from remain_2 (see which is smaller)
                        val_1 = remain_1[0]
                        val_2 = remain_2[0] + remain_2[1]

                        # return the sum-smaller value
                        return sum(nums)-min(val_1, val_2)
                    else:
                        return sum(nums)-remain_1[0]
                else:
                    if sum(multiples_of_3):
                        return sum(multiples_of_3)
                    else:
                        return 0
            
            elif remmainder == 2:
                
                # checking if there are any values if not return 0
                if len(remain_1)<2 or remain_2:
                    if len(remain_1)>=2:
                        # either remove 1 from remain_2 or 2 from remain_1 (see which is smaller)
                        val_1 = remain_2[0]
                        val_2 = remain_1[0] + remain_1[1]

                        # return the sum-smaller value
                        return sum(nums)-min(val_1, val_2)
                    else:
                        return sum(nums)-remain_2[0]
                else:
                    if sum(multiples_of_3):
                        return sum(multiples_of_3)
                    else:
                        return 0
```

```python
s = Solution()
# Input: 
nums = [1,1,3]

s.maxSumDivThree(nums)

# Output: 3
```

```python
s = Solution()
# Input: 
nums = [3,6,5,1,8]

s.maxSumDivThree(nums)

# Output: 18
# Explanation: Pick numbers 3, 6, 1 and 8 their sum is 18 (maximum sum divisible by 3).
```
