```python
# using the formula given

class Solution:
    def smallestRepunitDivByK(self, k: int) -> int:
        
        # for all values of n = 1, 11, 111, 111
        # it can never be able to do n%5 or n%2 to get a non 0 value
        if k%5==0 or k%2==0:
            return -1

        # storing the value rather than the len of the number
        target_num = k
        current_num = 0

        # running from 1 to 10 (a huge ass length)
        for i in range(1,1000000):

            # you dont need to divide teh oriignal number but can minus off a format instead
            # current length-1 *10 * 7
            calc_num = (current_num * 10+1)

            # checkinf if it is dividable by the target_number
            if calc_num%target_num==0:
                print(f'value found at {calc_num}')
                return i
            
            # reassigning the currenum to be the calc num
            current_num = calc_num
            
        # this means after the long ass loop it is unable to find a number
        return -1
```

```python
# Input: 
s = Solution()

k = 3
s.smallestRepunitDivByK(k)


# Output: 1
# Explanation: The smallest answer is n = 1, which has length 1.
```
