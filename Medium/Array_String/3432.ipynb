# 3432. Count Partitions with Even Sum Difference

```python
from typing import List
class Solution:
    def countPartitions(self, nums: List[int]) -> int:

        # create a pre calculated sum
        # moving from left to right
        sum_left_arr = []
        sum_right_arr = []
        
        sum = 0
        for values in nums:
            sum += values
            sum_left_arr.append(sum)

        for values in nums:
            sum_right_arr.append(sum)
            sum -= values

        answer = 0

        # calculating the difference
        for index in range(0, len(sum_left_arr)-1):

            if (sum_left_arr[index] - sum_right_arr[index+1])%2 != 0:
                continue
            
            else:
                answer+=1
            
        return answer
    
# time complexity of O(3*N)=O(N) because there are 3 for loops
# space complexity of O(2*N)=O(N) because 2 arrays are created
```

```python
# model answer

from typing import List
class Solution:
    def countPartitions(self, nums: List[int]) -> int:

        # the problem is to check SL - SR % 0
        # we know that SL + SR = ST too as well
        # thereofre the final equation can be simplified to (ST-SR)-SR % 0
        # == (ST - 2(SR)) % 0

        # the 2 SR part is always even because it is a multiple of 2....

        total_sum = sum(nums)

        if total_sum%2 == 0:
            return len(nums)-1
        
        else:
            return 0
```

```python
nums = [1, 2, 3]
sol = Solution()

sol.countPartitions(nums)
```
