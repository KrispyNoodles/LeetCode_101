# 1441. Build an Array With Stack Operations

### Problem  
You are given an integer array `target` and an integer `n`.

You have an initially empty stack and two operations:

- **"Push"** → push the next integer from the stream onto the stack  
- **"Pop"** → remove the top element from the stack  

You also have a stream of integers from `1` to `n` in order.

Use these operations to build the stack so that (from bottom to top) it becomes exactly equal to `target`. You must follow these rules:

1. If the stream still has numbers, read the next number and **Push** it.
2. You may **Pop** the top of the stack if needed.
3. Once the stack matches `target`, stop all operations immediately.
4. Return any valid list of operations.

---

### Example 1  
**Input:** `target = [1,3], n = 3` **Output:** `["Push","Push","Pop","Push"]`

**Explanation:**  
- Read 1 → Push → stack = [1]  
- Read 2 → Push → stack = [1,2]  
- Pop → stack = [1]  
- Read 3 → Push → stack = [1,3]  

---

### Example 2  
**Input:** `target = [1,2,3], n = 3` **Output:** `["Push","Push","Push"]`

---

### Example 3  
**Input:** `target = [1,2], n = 4` **Output:** `["Push","Push"]`

**Explanation:**  
Stop early because the stack already equals the target.

---

### Constraints  
- `1 <= target.length <= 100`  
- `1 <= n <= 100`  
- `1 <= target[i] <= n`  
- `target` is strictly increasing

```python
from typing import List

# My solution
class Solution:
    def buildArray(self, target: List[int], n: int) -> List[str]:

        final_array = []

        # n is how long your list can get (a super set of the targets)
        for i in range(1, n+1):
            
            # for situations where n is alot larger than the values in the target
            if i > max(target):
                return final_array
            
            # a push is always requried
            final_array.append("Push")
            
            # a pop is required if the value is not in the target
            if i not in target:
                final_array.append("Pop")

        return final_array
```

```python
s = Solution()

target = [1,3]
n = 3
# Output: ["Push","Push","Pop","Push"]

target = [1,2]
n = 4

# Output: ["Push","Push"]

s.buildArray(target, n)

```

```python
from typing import List

# My solution
class Solution:
    def buildArray(self, target: List[int], n: int) -> List[str]:

        final_array = []

        # position counter
        i = 0

        for num in target:

            # incrementing values using while
            while i < num - 1:

                # these are values that are added in then removed immediately after
                final_array.append("Push")
                final_array.append("Pop")

                # counter incremented
                i+=1
            
            # for values we want to keep
            final_array.append("Push")
            
            # counter incremented
            i+=1

        return final_array
```

```python
s = Solution()

# assume its alredy sorted because it is added incrementally

target = [1,3]
n = 3
# Output: ["Push","Push","Pop","Push"]

# target = [1,2]
# n = 4

# Output: ["Push","Push"]

s.buildArray(target, n)

```
