# 1930. Unique Length-3 Palindromic Subsequences

### Problem  
Given a string `s`, return the number of **unique** palindromic subsequences of **length 3**.

Notes:

- Even if the same subsequence can be formed in multiple ways, it is counted **only once**.
- A palindrome reads the same forward and backward.
- A subsequence is formed by deleting zero or more characters without changing relative order.

Example: `"ace"` is a subsequence of `"abcde"`.

---

### Example 1  
**Input:** `s = "aabca"` **Output:** `3`

**Explanation:** Valid palindromic subsequences are:  
- `"aba"`  
- `"aaa"`  
- `"aca"`

---

### Example 2  
**Input:** `s = "adc"` **Output:** `0`

---

### Example 3  
**Input:** `s = "bbcbaba"` **Output:** `4`

**Explanation:** Valid palindromic subsequences are:  
- `"bbb"`  
- `"bcb"`  
- `"bab"`  
- `"aba"`

---

### Constraints  
- `3 <= s.length <= 10^5`  
- `s` consists only of lowercase English letters

```python
# Approach 1: Count Letters In-Between

class Solution(object):
    def countPalindromicSubsequence(self, s):

        # finding all the unique values
        alphbet_set = set(s)
        answer = 0
        
        for alphabet in alphbet_set:

            # creating variables that finds the first and last of the value
            # index finds the index of the first while rindex finds the value of the last
            front, end = s.index(alphabet), s.rindex(alphabet)

            # print(front, end)
            
            # creating a splice of the array
            # s[start : stop] â†’ includes start, excludes stop
            inbetween_str = s[front+1:end]

            # finding all possible values that are within
            sum = len(set(inbetween_str))

            # adding all
            answer += sum


        return answer
```

```python
sol = Solution()

# return the number of unique palindromes of length three that are a subsequence of s.
# subsequence are characters that do not need to be sequential

s = "aabca"

sol.countPalindromicSubsequence(s)

# Output: 3
# Explanation: The 3 palindromic subsequences of length 3 are:
# - "aba" (subsequence of "aabca")
# - "aaa" (subsequence of "aabca")
# - "aca" (subsequence of "aabca")
```
