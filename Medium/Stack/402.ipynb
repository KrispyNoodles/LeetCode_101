```python
import sys

class Solution:
    def removeKdigits(self, num: str, k: int) -> str:
        
        # finding the next highest
        # decreasing monotonic stack

        num_stack = []

        for values in num:
            
            while num_stack and values < num_stack[-1] and k > 0:
                pop_val = num_stack.pop()

                # decrease k to negative
                k -= 1

            num_stack.append(values)

        # if k is not negative it means the values are increasing in the array
        # therefore remove the last few values from the array
        if k:
            num_stack = num_stack[:-k]

        # sys.set_int_max_str_digits(100000)
        # # convert to int to remove 0's in front and back to str as an answer
        # return str(int("".join(num_stack)))

        # alterantively
        result =  "".join(num_stack).lstrip('0')

        if result:
            return result
        else:
            return "0"
```

```python
# Input: 

num = "10"
k = 1

sol = Solution()
sol.removeKdigits(num, k)

# Output: "0"
```
