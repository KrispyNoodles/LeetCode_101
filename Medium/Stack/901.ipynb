```python
# After Understanding Stock Spanner
# it is an application that uses a decreasing monotonic stack
class StockSpanner:

    def __init__(self):

        # initialising an array
        self.imp_stack = []
        

    def next(self, price: int) -> int:

        if not self.imp_stack or self.imp_stack[-1][0] > price:

                # just append it
                self.imp_stack.append((price, 1))

                # return value of 1
                return 1
        
        else:
            # if the current price is higher than the one in the stack
            # repeat this loop with the value in the top of the stack
            days_counter = 1

            # equal prices will be counted as a continued high
            while self.imp_stack and price >= self.imp_stack[-1][0]:
                
                # remove all the useless old values
                _, days = self.imp_stack.pop()
                
                # adding all the days togther
                days_counter += days

            # append the new price
            self.imp_stack.append((price, days_counter))

            return days_counter
```

```python
# implenentation of a decreasing monotonic stack

# the proper and correct way
# no need to do for loop as the function being called many times is the for loop
class StockSpanner:

    def __init__(self):
        

        self.monotone_stack = []
        # syntax is (price, days)
              
    def next(self, price: int) -> int:

        # no need to reassign stack as this is a pointer created
        stack = self.monotone_stack
        days_counter = 1

        # if the current price is more than or equal to the value in the stack
        while stack and price >= stack[-1][0]:

            # pop it from the stack
            _, days = stack.pop()
            days_counter += days

        # append the current price and days_counter
        stack.append((price, days_counter))    
        
        return days_counter
```
