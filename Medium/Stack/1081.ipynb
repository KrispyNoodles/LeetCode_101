# 1081. Smallest Subsequence of Distinct Characters

```python
from collections import defaultdict

class Solution:
    def smallestSubsequence(self, s: str) -> str:

        # creating an increasing motonic stack

        # create a dict to check if there are any charcter that appears after
        s = list(s)

        count_dict = defaultdict(int)

        for char in s:
            count_dict[char] += 1

        # create increasing monotonic stack
        stack_char = []

        # set to ensure each value is only added once
        set_char = set()

        for char in s:
            # decreasing the count by 1 because it is viewed
            count_dict[char] -= 1

            # checking if it in the set before, else skip
            if char in set_char:
                continue

            # checking if the current character is smaller than the one within 
            # to optimise the increase of the stack?

            # Condition: char < stack_char[-1] is correct. This means the new character (char) is better (smaller) 
            # than the current top of the stack. We want to pop the larger element
            while stack_char and char < stack_char[-1]:
                
                # checkinf if the value appears later
                # if it is we can pop the value and take the later one
                if count_dict[stack_char[-1]] >= 1:
                    # pop from both
                    char_poped = stack_char.pop()
                    set_char.remove(char_poped)
                    
                # just exit the while loop
                else:
                    break
            
            stack_char.append(char)
            set_char.add(char)
        
        return "".join(stack_char)
        
# time complexity is O(n) because it skips character from the if char in set_char code
# if not it is usually O(nlog(n)) because of motonic stack
# space complexity of O(n)
```

```python
# Input: 

s = "cbacdcbc"
sol = Solution()

sol.smallestSubsequence(s)
# "acdb"
```

```python
# Input: 
s = "bcabc"
sol = Solution()

sol.smallestSubsequence(s)

# Output: "abc"
```
