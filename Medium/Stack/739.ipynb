# 739. Daily Temperatures

# MY SOLN

```python
from typing import List

# my soln easy!!
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        
        # increasing monotonic stack

        # stack containing the indexes
        val_stack = []
        value_array = [0]*len(temperatures)

    
        nu_of_days = [0]*len(temperatures)

        for index in range(len(temperatures)):

            # if the val_stack exist
            # and the current value is larger than the one in the stack
            # it has to be >= so that the next temp is defo hotter
            while val_stack and temperatures[index] > temperatures[val_stack[-1]]:
                
                # remove the index
                pop_index = val_stack.pop()

                value_array[pop_index] = temperatures[index]

                # number of days to wait is the pop_index - current index
                nu_of_days[pop_index] = index-pop_index

            val_stack.append(index)
        
        # dont need to work with the val_stack indexes
        # as theses are the values that are largest and therefore unable to find the next highest
        # therefore that have 0 days till it hits the warmer temp
        return nu_of_days
```

```python
# Input: 
temperatures = [73,74,75,71,69,72,76,73]

s = Solution()
s.dailyTemperatures(temperatures)

# Output: [1,1,4,2,1,1,0,0]
```
