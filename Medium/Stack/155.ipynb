# 155. Min Stack


### Problem  
Design a stack that supports the following operations in **O(1)** time:

- `push(val)` — push an element onto the stack  
- `pop()` — remove the top element  
- `top()` — return the top element  
- `getMin()` — return the **minimum element** in the stack  

Implement the `MinStack` class:

- **MinStack()** — initializes the stack  
- **void push(int val)**  
- **void pop()**  
- **int top()**  
- **int getMin()**

---

### Example 1  
**Input:**  
`["MinStack","push","push","push","getMin","pop","top","getMin"]`  
`[[],[-2],[0],[-3],[],[],[],[]]`  
**Output:**  
`[null,null,null,null,-3,null,0,-2]`

**Explanation:**  
```
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
```

---

### Constraints  
- `-2^31 <= val <= 2^31 - 1`  
- `pop`, `top`, and `getMin` will always be called on **non-empty** stacks  
- Up to `3 * 10^4` operations will be performed

````python
# creating another stack for the minium value
# precomputing the min value
class MinStack:

    def __init__(self):
        self.array = []

        # it does not store an order from small to big
        # but rather a small version of the main stack
        self.min_stack = []

    # adding the elemtn into the array
    def push(self, val: int) -> None:
        self.array.append(val)

        # if there is no value, the current value is the smallest
        if not self.min_stack:
            self.min_stack.append(val)
        
        else:
            # the current smallest is smaller than that value
            if self.min_stack[-1] >= val:
                self.min_stack.append(val)         

            else:
                self.min_stack.append(self.min_stack[-1])     

    # removing the top elemnt
    def pop(self) -> None:
        self.array.pop()
        self.min_stack.pop()

    # retrieving the top elemtn
    def top(self) -> int:
        return self.array[-1] 

    # returnng the smallest value in the array
    def getMin(self) -> int:
        return self.min_stack[-1]
        
````
