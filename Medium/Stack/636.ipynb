# 636. Exclusive Time of Functions

### Problem  
You are running a program with `n` functions on a **single-threaded CPU**.  
Each function has a unique ID from `0` to `n-1`.

Execution uses a **call stack**:

- When a function **starts**, its ID is pushed on the stack  
- When it **ends**, its ID is popped  
- The function at the **top of the stack** is currently running  
- Logs record each function start/end with timestamps  

Logs are given in the format:

```
"{function_id}:{start|end}:{timestamp}"
```

Example:  
- `"0:start:3"` → function `0` starts at beginning of time `3`  
- `"1:end:2"` → function `1` ends at end of time `2`

A function’s **exclusive time** = total time it spent executing (excluding time spent in child calls).

Return an array where `result[i]` is the exclusive time of function `i`.

---

### Example 1  
**Input:** `n = 2`, `logs = ["0:start:0","1:start:2","1:end:5","0:end:6"]`  
**Output:** `[3,4]`

**Explanation:**  
- Function 0 runs from time 0 → 1 (2 units)  
- Function 1 runs from time 2 → 5 (4 units)  
- Function 0 resumes at time 6 (1 unit)  
Total: `2 + 1 = 3` for function 0, `4` for function 1

---

### Example 2  
**Input:**  
`n = 1`,  
`logs = ["0:start:0","0:start:2","0:end:5","0:start:6","0:end:6","0:end:7"]`  
**Output:** `[8]`

---

### Example 3  
**Input:**  
`n = 2`,  
`logs = ["0:start:0","0:start:2","0:end:5","1:start:6","1:end:6","0:end:7"]`  
**Output:** `[7,1]`

---

### Constraints  
- `1 <= n <= 10^4`  
- `1 <= logs.length <= 10^4`  
- Logs are valid and timestamps are non-decreasing

````python
from typing import List

# Second attempt
class Solution:
    def exclusiveTime(self, n: int, logs: List[str]) -> List[int]:

        # to get the unique number of functions
        final_ans = [0]*n

        # checks which function is currently running atm
        stack = []

        # initatilisation
        time_run = 0

        # retrieving a formatted array
        for string in logs:

            # the time include is found to determine the start and end, therefore the start and end flags are kinda useless?
            function, start_end, time = string.split(':')

            # converting time into int
            time = int(time)

            # converting function into int so that it can be used an index
            function = int(function)
            
            # a function starting
            if start_end == 'start':

                # if the stack is empty, there is no time to include into the answer
                if stack != []:
                    # add the time that was ran to the previous element first before removing it
                    # time calculated is the previous current time minus the previous time(time_run)
                    # it has to be plus equal to add all the time it has ran before
                    final_ans[stack[-1]] += time - time_run
                    
                # including the current function
                stack.append(function)

                # start the time counter of the current fn
                time_run = time

            # if a function ends
            elif start_end == 'end':

                # adding the time that was running (+1 because end counts +1)
                # current time - previous time + 1
                # += for the same reasoning above
                final_ans[stack[-1]] += (time - time_run) + 1

                # updating the time_run to be the current time
                # + 1 to include the ending taking up 1 run time
                time_run = time + 1
                
                # removing the previous function
                stack.pop()                           
   
        return final_ans
````

````python
from typing import List

# Double stack - Super clean and efficient
class Solution:
    def exclusiveTime(self, n: int, logs: List[str]) -> List[int]:

        # to get the unique number of functions
        final_ans = [0]*n

        # checks the time of the current function
        call_stack = []

        # initatilisation
        idle_duration = []

        # this understanding is because the current time minus the number of time it has not worked
        # execution_time = time - idle_duration

        # retrieving a formatted array
        for string in logs:

            # the time include is found to determine the start and end, therefore the start and end flags are kinda useless?
            function, start_end, time = string.split(':')

            # converting into int
            time = int(time)
            function = int(function)
            
            # a function starting
            if start_end == 'start':
                    
                # including the current function
                call_stack.append(time)

                # there is no idle_duration for something that is starting
                idle_duration.append(0)

            # if a function ends
            elif start_end == 'end':

                # retrieving previous function timee
                prev_time = call_stack.pop()

                # calculating duration that it was queued for
                # + 1 because its at the end
                duration = time - prev_time +1

                # retrieving previous idle_duration
                idle = idle_duration.pop()

                # assigning the answer
                final_ans[function] = duration - idle

                # if there is a function being run, add the time to the previous element
                if call_stack:
                    idle_duration[-1] += duration
   
        return final_ans
````

````python
from typing import List

# Second attempt
class Solution:
    def exclusiveTime(self, n: int, logs: List[str]) -> List[int]:

        # this stack is to queue the functions
        function_stack = []

        calc_time = 0

        # an array of 0
        final_ans = [0]*int(n)

        for log in logs:

            # splitting based of ":"
            function, start_stop, time = log.split(':')

            # converting function and time to be ints
            function = int(function)
            time = int(time)

            if start_stop == 'start':

                # if stack is not empty
                if function_stack:

                    time_ran = time - calc_time

                    # add the time added in to the prev dude
                    final_ans[function_stack[-1]] += time_ran

                # change the calc time to be now
                calc_time = time

                # welcome to the queue
                # append the function that is being queued
                function_stack.append(function)

            elif start_stop == 'end':    

                # previous one that was running for the prev function
                # current_time - calc_time
                # + 1 because its the end
                time_ran = time - calc_time + 1

                # remove the function from the top
                final_ans[function_stack.pop()] += time_ran

                # reset the calc_time to be current time + 1 (because of 1 sec to end too)
                # then it can count for the next function in the stack
                # After a function ends at time t, the CPU moves to time t+1.
                calc_time = time + 1
            
        return final_ans
````

````python
s = Solution()

# Input: 
n = 2
logs = ["0:start:0","1:start:2","1:end:5","0:end:6"]

# 0 0 1 0 1 0 
# add 0
# queue stack
# [0]

# add 1
# queue stack
# [0, 1]

# stop 1
# queue stack
# [0]

# Output: [3,4]

s.exclusiveTime(n, logs)
````
