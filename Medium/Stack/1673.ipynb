# 1673. Find the Most Competitive Subsequence

```python
# need to redo because had help on the formula
from typing import List

class Solution:
    def mostCompetitive(self, nums: List[int], k: int) -> List[int]:
        
        # the solution is to create an increasing monotonic stack
        # and then return the stack as the answer

        stack_num = []


        for index in range(len(nums)):

            while stack_num and nums[index] < stack_num[-1]:

                # before I pop, I need to check if there is enough values that can be added
                # new stack length - 1 (from popping) + the total length-the current index is more than the value k
                if len(stack_num)-1 + (len(nums)-index) >= k:
                    stack_num.pop()
                else:
                    break
                
            stack_num.append(nums[index])

        # return the first value of k in case the last value is included
        return stack_num[:k]
    
# time complecity of O(n) for looping through all the elemtns at least once in the array 
# space complexity of O(n)
```

```python
nums = [2,4,3,3,5,4,9,6]
k = 4

sol = Solution()
sol.mostCompetitive(nums, k)

# [2,3,3,4]
```

```python
nums = [5,4,3,2,1]
k = 3

sol = Solution()
sol.mostCompetitive(nums, k)

# answer [3,2,1]
```
