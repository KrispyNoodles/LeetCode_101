```python
from collections import defaultdict

class Solution:
    def removeDuplicateLetters(self, s: str) -> str:
        
        # do a increasing monotonic stack sequentially on the different alphaebts

        # converting alphabets into values
        # there is no need to conver alphabets to numbers because python already agrees that a is the smallest character
        
        # a set to check if this character has been visited before
        visited_char = set()

        # montonic stack
        char_stack = []

        visisted_dict = defaultdict(int)

        # creating a dict to count the total values of each appearance
        for character in s:
            visisted_dict[character] +=1


        for index in range(len(s)):

            # subtract by 1 after it has been viewed in the looop
            visisted_dict[s[index]] -= 1

            # if the character has been seen before skip it
            if s[index] in visited_char:
                continue
            
            # checking if the current value is larger than the one inside
            # I am trying to create an increasing montonic stack since (a is the smallest...)

            # If the new character coming in is smaller than the character currently at the top of my increasing stack..
            # then we pop the top one because the new, smaller one is better and maintains our increasing structure more optimally
            while char_stack and s[index] < char_stack[-1]:
                
                # this migth causes O(n) * O(n) = O(n^2)
                # one from the while loop and the other from creating a new array
                # and since it is inside a for loop, potentially O(n^3)
                # if char_stack[-1] in s[index+1:]:

                # using a dict instead
                if visisted_dict[char_stack[-1]]>=1:

                    # pop the stack and the set
                    char_removed = char_stack.pop()
                    visited_char.remove(char_removed)

                # if not it runs into an infinite loop
                else:
                    break 

            char_stack.append(s[index])

            # add the value into the visited_set
            visited_char.add(s[index])

        return "".join(char_stack)
```

```python
s = "bbcaac"
sol = Solution()

sol.removeDuplicateLetters(s)

# Output: "bac"
```
