```python
from typing import List

# my soln
class Solution:
    def nextGreaterElements(self, nums: List[int]) -> List[int]:

        # length of original array
        length = len(nums)

        # extending the list by twice
        nums = nums + nums
        
        # creating increasing monotonic stack
        # stack contianing index
        stack_val = []

        # array holding the ans
        temp_arr = [0]*len(nums)

        for index in range(len(nums)):

            # checking if stack_val exits
            # comparing the last element in the stack bigger than the current elemetns
            while stack_val and nums[index] > nums[stack_val[-1]]:

                pop_index = stack_val.pop()
                
                temp_arr[pop_index] = nums[index]

            stack_val.append(index)
        
        # index remaining
        for i in stack_val:

            # if it is the max_value assign -1
            # else assign the max number to it instead
            if nums[i] == max(nums):
                temp_arr[i] = -1

        return temp_arr[:length]
```

```python
sol = Solution()

# Input: 
nums = [1,2,3,2,1]
sol.nextGreaterElements(nums)

# Output: [2,3,-1,3,2]
```

```python
from typing import List

# they change the index but run twice the total len
# 22ms (Beats 65.97%) NOOB SOLUTION because it runs the double length

class Solution:
    def nextGreaterElements(self, nums: List[int]) -> List[int]:
        
        # length of original array
        length = len(nums)
        
        # creating increasing monotonic stack
        # stack contianing index
        stack_val = []

        # array holding the ans
        temp_arr = [-1]*len(nums)

        for index in range(len(nums)*2):

            # checking if stack_val exits
            # comparing the last element in the stack bigger than the current elemetns
            while stack_val and nums[index%length] > nums[stack_val[-1]]:

                pop_index = stack_val.pop()
                
                temp_arr[pop_index] = nums[index%length]

            if index < length:
                stack_val.append(index)
            
        return temp_arr
```

```python
sol = Solution()

# Input: 
nums = [1,2,3,2,1]
sol.nextGreaterElements(nums)

# Output: [2,3,-1,3,2]
```
