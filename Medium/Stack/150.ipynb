```python
class Solution(object):
    def evalRPN(self, tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """

        my_array = []

        for i in tokens:

            if i == "+":
                # print("plus is found")
                
                # operate the operate on the previous two values in my_array
                val_1 = my_array.pop()
                val_2 = my_array.pop()

                my_array.append(val_1 + val_2)

            elif i == "-":
                print("minus is found")
                
                # operate the operate on the previous two values in my_array
                val_1 = my_array.pop()
                val_2 = my_array.pop()

                my_array.append(val_2 - val_1)

            elif i == "*":
                # print("times is found")

                val_1 = my_array.pop()
                val_2 = my_array.pop()

                my_array.append(val_1 * val_2)

            
            elif i == "/":
                # print("divide is found")
                
                # since val_2 enter first it is the value that divides val_1
                val_1 = my_array.pop()
                val_2 = my_array.pop()
                
                # The division between two integers always truncates toward zero.
                # means 2.5 becomes 2

                # this causing the issue of 12 in LC and 22 in Vs Code
                # because of python version of python2 in LC vs python3 in Vs Code

                my_array.append(int(val_2/ val_1))
                
            else:
                my_array.append(int(i))

        return my_array.pop()
        
```

```python
s = Solution()

# tokens = ["2","1","+","3","*"]
# # Output: 9

# tokens = ["4","13","5","/","+"]
# # Output: 6


tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
# Output: 22

s.evalRPN(tokens)
```
