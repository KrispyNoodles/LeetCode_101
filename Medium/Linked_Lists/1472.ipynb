```python
# Definition for douvly-linked list.
class ListNode:
    def __init__(self, url="", next=None, prev=None):
        self.url = url
        self.next = next
        self.prev = prev

class BrowserHistory:

    def __init__(self, homepage: str):

        # create a current to track where you are at
        self.current = ListNode(homepage)
        
    # adding a new node in (CORRECT)
    def visit(self, url: str) -> None:

        # add a new node to the current
        # create a new node
        node_added = ListNode(url)

        self.current.next = node_added
        node_added.prev = self.current

        self.current = node_added
    
    # only can move certain steps of value
    def back(self, steps: int) -> str:

        count = 0

        # current is where you are at
        curr = self.current

        # it has to be curr.next != none to return the last link
        # if not it will return the none
        while curr != None and count < steps:

            # vairable to catch second_last
            second_last = curr

            curr = curr.prev

            # increment count
            count += 1

        # checkinf if the curr is none
        if curr is None:
            # take the second last eelment
            curr = second_last

        # updating the current
        self.current = curr

        return curr.url

    def forward(self, steps: int) -> str:

        count = 0

        # current is where you are at
        curr = self.current

        while curr != None and count < steps:
            
            # vairable to catch second_last
            second_last = curr

            curr = curr.next

            # increment count
            count += 1
        
        # checkinf if the curr is none
        if curr is None:
            # take the second last eelment
            curr = second_last

        # updating the current
        self.current = curr

        return curr.url
        
# Your BrowserHistory object will be instantiated and called as such:
# obj = BrowserHistory(homepage)
# obj.visit(url)
# param_2 = obj.back(steps)
# param_3 = obj.forward(steps)
```
